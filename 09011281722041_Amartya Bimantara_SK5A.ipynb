### PENENTUAN PRESTASI BELAJAR SISWA MENGGUNAKAN METODE FUZZY MAMDANI

Prestasi belajar siswa perlu diperhatikan agar dapat dievaluasi. Untuk itu Dibutuhkan sebuah metode logika fuzzy untuk penentuan prestasi belajar siswa. Prestasi belajar siswa akan bernilai baik atau buruk diukur dengan menggunakan 3 variabel yaitu gaya belajar visual, gaya belajar auditorial, dan gaya belajar kinestetik.

##### Import library:

```python
import numpy as np
import skfuzzy as fuzz
import matplotlib.pyplot as plt
from skfuzzy import control as ctrl
```

Input : 
Gaya belajar visual dibagi menjadi 3 bagian yakni:
- Rendah (1 - 50)
- Sedang (50 - 70)
- Tinggi (40 - 75)

Gaya belajar auditorial dibagi menjadi 3 bagian yakni:
- Rendah (1 - 50)
- Sedang (50 - 70)
- Tinggi (40 - 75)

Gaya belajar kinestetik dibagi menjadi 3 bagian yakni:
- Rendah (1 - 50)
- Sedang (50 - 70)
- Tinggi (40 - 75)

Output :
Prestasi belajar yang dibagi menjadi 2 bagian yakni:
- Buruk (1 - 20)
- Baik (50 - 80)

# Fuzzifikasi

Proses untuk mengubah variabel non numerik menjadi variabel non fuzzy (numerik) menjadi variabel fuzzy (linguistik).

## Fuzzifikasi Gaya Belajar Visual

```python
visual = ctrl.Antecedent(np.arange(1,101,1),'visual')
visual_x = np.arange(1,101,1) #untuk plotting saja
```

Gaya belajar visual memiliki rentan nilai dari 1 - 75

Untuk gaya belajar visual, fungsi keanggotaan dibagi menjadi 3 bagian:
$$ Gaya Belajar Visual $$
$$ Rendah = 1 - 50 $$
$$ Sedang = 50 - 70 $$
$$ Tinggi = 40 - 75 $$


Dengan nilai linguistik gaya belajar visual

\begin{equation}
  visual['rendah'] = \mu VisualRendah(x)=\begin{cases}
    1, & \text{$x\leq50$}.\\
    (60 - x)/(60 - 50) , & \text{$50 \leq x \leq 60$}.\\
    0, & \text{$x\geq 60$}.
  \end{cases}
  \\
  visual['sedang'] = \mu VisualSedang(x)=\begin{cases}
    0, & \text{$x\leq 50$ atau $x\geq75$}.\\
    (x - 50)/(60 - 50) , & \text{$50 \leq x \leq 60$}.\\
    (70 - x)/(70 - 60) , & \text{$60 \leq x \leq 70$}.\\
  \end{cases}
  \\
 visual['tinggi'] = \mu VisualTinggi(x)=\begin{cases}
    0, & \text{$x\leq 60$}.\\
    (x - 60)/(75 - 60) , & \text{$60 \leq x \leq 75$}.\\
    1, & \text{$x\geq 75$}.
  \end{cases}
\end{equation}


```python
visual['rendah'] = fuzz.trapmf(visual.universe, [1, 1, 50, 60])
visual['sedang'] = fuzz.trimf(visual.universe, [50, 60, 70])
visual['tinggi'] = fuzz.trapmf(visual.universe, [60, 75, 100, 100])
```

## Plotting Membership Function Visual

Plotting fungsi keanggotaan $\mu VisualRendah(x)$

```python
visual_rendah = fuzz.trapmf(visual_x,[1,1,50,60])
fig, (ax0) = plt.subplots(nrows=1,figsize=(6,6))
ax0.plot(visual_x,visual_rendah,'blue')
ax0.set_title('Visual Rendah')
```

Plot fungsi keanggotaan $\mu VisualSedang(x)$

```python
visual_sedang = fuzz.trimf(visual_x,[50,60,70])
fig2, (ax1) = plt.subplots(nrows=1,figsize=(6,6))
ax1.plot(visual_x,visual_sedang,'orange')
ax1.set_title('Visual Sedang')
```

Plot fungsi keanggotaan $\mu VisualTinggi(x)$

```python
visual_tinggi = fuzz.trapmf(visual_x,[60,75,100,100])
fig, (ax0) = plt.subplots(nrows=1,figsize=(6,6))
ax0.plot(visual_x,visual_tinggi,'green')
ax0.set_title('Visual Tinggi')
```

```python
visual.view()
```

# Evaluasi Rule

Pada tahap ini, membuat aturan-aturan fuzzy yang dinyatakan dalam bentuk IF .... THEN ....

Pada sistem ini berdasarkan jumlah fuzzifikasi dari variabel visual yang berjumlah 3, fuzzifikasi dari variabel auditorial yang berjumlah 3, dan fuzzifikasi dari variabel kinestetik yang berjumlah 3, maka rule yang dibuat berjumlah 3 x 3 x 3 =  27 rule

1. IF Visual Rendah AND Auditorial Rendah AND Kinestetik Rendah THEN PrestasiBelajar Buruk

18. IF Visual Sedang AND Auditorial Tinggi AND Kinestetik Tinggi THEN PrestasiBelajar Baik

### Rumus

- Persegi panjang: A = p x l

- Segitiga : A = 1/2 (p x l)

$$ A_1 = \frac{ p \times l }
{ 2 }$$

$$ A_1 = 5 $$

Mencari $ A_2 $ :


$$ A_2 = { p\times l }$$

$$ A_2 = { 60 \times 0.5}$$

$$ A_2 = 30 $$

Untuk mencari nilai X, gunakan rumus:

- Segitiga: X = a + 2/3( b - a )

- Persegi panjang: X = 1/2 (a + b)

dimana a adalah batas atas dan b adalah batas bawah

$$ X_1 = { a + 2/3 ( b - a ) } $$
