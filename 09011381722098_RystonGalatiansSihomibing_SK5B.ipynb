# APLIKASI LOGIKA FUZZY UNTUK PERAMALAN BEBAN LISTRIK JANGKA PENDEK

```python
import numpy as np #Numpy adalah modul yang menyediakan objek matematika yang melakukan perhitungan.
import skfuzzy as fuzz #Skfuzzy berfungsi agar fuzzy bisa digunakan
from skfuzzy import control as ctrl #Agar ctrl bisa digunakan
```

Pada input pertama dimasukkan perintah import numpy as np, library numpy digunakan untuk manipulasi data matriks sehingga ini mempermudah proses komputasi datanya. Sedangkan library skfuzz digunakan untuk implementasi komputasi fuzzy. Ada beberapa hal yang perlu diketahui dalam memahami sistem logika fuzzy yaitu:
1. Variabel fuzzy :
   Merupakan variabel yang hendak dibahas dalam suatu sistem fuzzy.
2. Himpunan fuzzy :
   Merupakan suatu grup yang mewakili suatu kondisi atau keadaan tertentu dalam suatu variabelfuzzy.
3. Semesta pembicaraan :
   Adalah keseluruhan nilai yang diperbolehkan untuk dioperasikan dalam suatu variabel fuzzy.
4. Domain : 
   Keseluruhan nilai yang di izinkan dalam semesta pembicaraan dan boleh di operasikan dalam suatu 
   himpunan fuzzy.

Representasi Linier Pada representasi linier, pemetaan input kederajat keanggotaannya digambarkan sebagai sebuah garis lurus. Bentuk ini paling sederhana dan menjadi pilihan yang baik untuk mendekati suatu konsep yang kurang jelas.
Representasi Kurva Segitiga Kurva segitiga pada dasarnya merupakan gabungan antara 2 garis (linear). Nilai-nilai disekitar b memiliki derajat keanggotaan turun cukup tajam (menjauhi 1).
Representasi Kurva Bentuk Bahu Daerah yang terletak di tengah-tengah suatu variabel yang direpresentasikan dalam bentuk segitiga, pada sisi kanan dan kirinya akan naik turun. Tetapi terkadang salah satu sisi dari variabel tersebut tidak mengalami perubahan. Himpunan fuzzy “bahu”, bukan segitiga, digunakan untuk mengakhiri variabel suatu daerah fuzzy.


## 1. Fuzzyfikasi
Fuzzyfikasi merupakan proses untuk mengubah input sistem yang mempunyai nilai tegas menjadi variabel linguistik munggunakan fungsi keanggotaan yang disimpan dalam basis pengetahuan fuzzy. Dalam hal ini, terdapat 5 variabel, yaitu 4 variabel input dan 1 variabel output.
 
## Variabel Input  (Antecedents) :
<br>1. Beban Historis : Tinggi dan Rendah.
<br>2. Isi dan Bahan : Rendah, Sedang, Tinggi.
<br>
## Variabel Output (Consequents) :
Beban Ramal
## Nilai Lingustik
Kemudian terdapat pula nilai lingustik atau semesta pembicaraan yang merupakan keseluruhan nilai yang diperbolehkan untuk dioperasikan dalam suatu variabel fuzzy, dan terdapat juga Domain fuzzy yang merupakan keseluruhan nilai yang diizinkan dalam semesta pembicaraan dan boleh dioperasikan dalam suatu himpunan fuzzy. 

```python
bh = ctrl.Antecedent(np.arange(0,400,1),'beban historis') #Inputnya
sh = ctrl.Antecedent(np.arange(22,32,1),'suhu') #Inputnya
br = ctrl.Consequent(np.arange(0,400,1),'beban ramal') #Outputnya
```

```python
bh['min']=fuzz.trimf(bh.universe,[0,30,52]) #Untuk membuat grafik segitiga
bh['vs']=fuzz.trimf(bh.universe,[52,90,120]) #Untuk membuat grafik segitiga
bh['s']=fuzz.trimf(bh.universe,[120,150,180]) #Untuk membuat grafik segitiga
bh['m']=fuzz.trimf(bh.universe,[180,220,240]) #Untuk membuat grafik segitiga
bh['b']=fuzz.trimf(bh.universe,[240,270,300]) #Untuk membuat grafik segitiga
bh['vb']=fuzz.trimf(bh.universe,[300,330,360]) #Untuk membuat grafik segitiga
bh['max']=fuzz.trimf(bh.universe,[360,390,430]) #Untuk membuat grafik segitiga
bh.view() #Menampilkan hasil
```

```python
sh['rendah']=fuzz.trimf(sh.universe,[0,22,27]) #Untuk membuat grafik segitiga
sh['rata - rata']=fuzz.trimf(sh.universe,[22,27,32]) #Untuk membuat grafik segitiga
sh['tinggi']=fuzz.trimf(sh.universe,[27,32,32]) #Untuk membuat grafik segitiga
sh.view() #Untuk menampilkan hasil
```

```python
br['min']=fuzz.trimf(bh.universe,[0,30,52]) #Untuk membuat grafik segitiga
br['vs']=fuzz.trimf(bh.universe,[52,90,120]) #Untuk membuat grafik segitiga
br['s']=fuzz.trimf(bh.universe,[120,150,180]) #Untuk membuat grafik segitiga
br['m']=fuzz.trimf(bh.universe,[180,220,240]) #Untuk membuat grafik segitiga
br['b']=fuzz.trimf(bh.universe,[240,270,300]) #Untuk membuat grafik segitiga
br['vb']=fuzz.trimf(bh.universe,[300,330,360]) #Untuk membuat grafik segitiga
br['max']=fuzz.trimf(bh.universe,[360,390,430]) #Untuk membuat grafik segitiga
br.view() #Untuk menampilkan hasil
```

## 2. Rule

```python
rule1=ctrl.Rule(bh['min']&sh['rendah'],br['min']) 
rule2=ctrl.Rule(bh['min']&sh['rata - rata'],br['min'])
rule3=ctrl.Rule(bh['min']&sh['tinggi'],br['min'])
rule4=ctrl.Rule(bh['vs']&sh['rendah'],br['vs'])
rule5=ctrl.Rule(bh['vs']&sh['rata - rata'],br['vs'])
rule6=ctrl.Rule(bh['vs']&sh['tinggi'],br['vs'])
rule7=ctrl.Rule(bh['s']&sh['rendah'],br['s'])
rule8=ctrl.Rule(bh['s']&sh['rata - rata'],br['s'])
rule9=ctrl.Rule(bh['s']&sh['tinggi'],br['s'])
rule10=ctrl.Rule(bh['m']&sh['rendah'],br['m'])
rule11=ctrl.Rule(bh['m']&sh['rata - rata'],br['m'])
rule12=ctrl.Rule(bh['m']&sh['tinggi'],br['m'])
rule13=ctrl.Rule(bh['b']&sh['rendah'],br['b'])
rule14=ctrl.Rule(bh['b']&sh['rata - rata'],br['b'])
rule15=ctrl.Rule(bh['b']&sh['tinggi'],br['b'])
rule16=ctrl.Rule(bh['vb']&sh['rendah'],br['vb'])
rule17=ctrl.Rule(bh['vb']&sh['rata - rata'],br['vb'])
rule18=ctrl.Rule(bh['vb']&sh['tinggi'],br['vb'])
rule19=ctrl.Rule(bh['max']&sh['rendah'],br['max'])
rule20=ctrl.Rule(bh['max']&sh['rata - rata'],br['max'])
rule21=ctrl.Rule(bh['max']&sh['tinggi'],br['max'])

```

```python
rule1.view()
rule2.view()
rule3.view()
rule4.view()
rule5.view()
rule6.view()
rule7.view()
rule8.view()
rule9.view()
rule10.view()
rule11.view()
rule12.view()
rule13.view()
rule14.view()
rule15.view()
rule16.view()
rule17.view()
rule18.view()
rule19.view()
rule20.view()
rule21.view()
```

# 3. Mesin Inferensi

Pada tahap mesin inferensi, dalam menentukan nominal beasiswa yang diberikan di Fakultas Teknik Universitas Majalengka akan menerapkan fungsi MIN untuk setiap rule (aturan) yang digunakan pada aplikasi fungsi implikasinya.

```python
oto1=ctrl.ControlSystem([rule1,rule2,rule3,rule4,rule5,rule6,rule7,rule8,rule9,rule10,rule11,rule12,rule13,rule14,rule15,rule16,rule17,rule18,rule19,rule20,rule21])
```

```python
oto=ctrl.ControlSystemSimulation(oto1)
```

```python
oto.input['beban historis']=60 #Input data simulasi
oto.input['suhu']=125 #Input data simulasi
```

```python
oto.compute() #Menampilkan data
print(oto.output['beban ramal'])
```

```python
br.view(sim=oto) #Menampilkan data
```

```python
oto.input['beban historis'] = 3.1
oto.input['suhu'] = 5.1
```

```python
oto.compute()
```

```python
print(oto.output['beban ramal'])
br.view(sim=oto)
```
