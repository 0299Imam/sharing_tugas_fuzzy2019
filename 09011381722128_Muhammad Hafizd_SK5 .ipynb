# Fuzzifikasi
Terdapat 2 Input dan 1 Output
Input = 
        1.Penghasilan
        2. pengeluaran
output = 1. angsuran


```python
penghasilan = ctrl.Antecedent(np.arange(300,3000, 10), 'penghasilan')
pengeluaran = ctrl.Antecedent(np.arange(75, 750, 10), 'pengeluaran')
angsuran = ctrl.Consequent(np.arange(100, 501, 10), 'angsuran')
```

Inputnya yang pertama adalah Penghasilan. dengan 3 fungsi keanggotaan. Yaitu : Low, Medium, High

```python
penghasilan['low'] = fuzz.trapmf(penghasilan.universe, [300, 300, 500, 1650])
penghasilan['medium'] = fuzz.trimf(penghasilan.universe, [500, 1650, 2800])
penghasilan['high'] = fuzz.trapmf(penghasilan.universe, [1650, 2800, 3000, 3000])
```

- Penghasilan
$$ Penhasilan [low] = 
\begin{cases}
    1 & ; x \leq 300\\
    \frac{(x-300)}{(3000-300)} & ;300 \leq x \leq\ 3000 \\
    0 & ;x \geq\ 3000 \\
\end{cases}
$$

- Penghasilan
$$ penghasilan [Medium] = 
\begin{cases}
    0 & ; x \leq 300\\
    \frac{(x-300)}{(1650-300)} & ;300 \leq x \leq\ 1650 \\
    \frac{(3000-x)}{(3000-1650)} & ;1650 \leq x \leq\ 3000 \\
\end{cases}
$$

- Penghasilan
$$ Penghasilan [High] = 
\begin{cases}
    0 & ; x \leq 300\\
    \frac{(3000-x)}{(3000-300)}\\
    1 & ;x \geq\ 3000 \\
\end{cases}
$$


# Mendeklarasikan Rule Base IF THEN


```python
rule1 = ctrl.Rule(penghasilan['high'] & pengeluaran['high'], angsuran['low'])
rule2 = ctrl.Rule(penghasilan['high'] & pengeluaran['medium'], angsuran['medium'])
rule3 = ctrl.Rule(penghasilan['high'] & pengeluaran['low'], angsuran['high'])
rule4 = ctrl.Rule(penghasilan['medium'] & pengeluaran['high'], angsuran['low'])
rule5 = ctrl.Rule(penghasilan['medium'] & pengeluaran['medium'], angsuran['low'])
rule6 = ctrl.Rule(penghasilan['medium'] & pengeluaran['low'], angsuran['low'])
rule7 = ctrl.Rule(penghasilan['low'] & pengeluaran['high'], angsuran['low'])
rule8 = ctrl.Rule(penghasilan['low'] & pengeluaran['medium'], angsuran['low'])
rule9 = ctrl.Rule(penghasilan['low'] & pengeluaran['low'], angsuran['low'])
```

```python
angsuran_ctrl = ctrl.ControlSystem(
    [rule1, rule2, rule3, rule4, rule5, rule6, rule7, rule8, rule9])
```

# Defuzzifikasi

```python
braking = ctrl.ControlSystemSimulation(angsuran_ctrl)

braking.input['penghasilan'] = 3100
braking.input['pengeluaran'] = 500

braking.compute()

print(braking.output['angsuran'])

angsuran.view(sim=braking)

plt.show()
```
