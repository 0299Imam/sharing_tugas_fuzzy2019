## 1. FUZZYFIKASI
Fuzzyfikasi merupakan proses untuk mengubah input sistem yang mempunyai nilai tegas menjadi variabel linguistik munggunakan fungsi keanggotaan yang disimpan dalam basis pengetahuan fuzzy. Dalam hal ini, terdapat 3 variabel, yaitu 2 variabel input dan 1 variabel output.

### Variabel Input  (Antecedents) : 
<br>1. SUHU : Dingin, Hangat, dan Panas.
<br>2. KELEMBAPAN : Kering, Normal, dan Basah.

### Variabel Output (Consequents) :
DURASI POMPA AIR : Cepat, Sedang, dan Lama

```python
x_suhu       = ctr.Antecedent(np.arange(0, 61, 1),'x_suhu')#range yang digunakan dalam kasus ini terhadap sensor suhu adalah 0-50 dengan loncatan 1
x_kelembapan = ctr.Antecedent(np.arange(0, 101, 1),'x_kelembapan')#range yang digunakn dalam kasus ini terhadap sensor kelembapan adalah 0-100 dengan loncatan 1
x_durasi     = ctr.Consequent(np.arange(0, 250, 1),'x_durasi')#range yang digunakan yaitu 0-250
```

```python
#inputan suhu
x_suhu['dingin'] = fuzz.trapmf(x_suhu.universe,(0,0,20,30)) #rentang nilai batas pada inputan suhu dingin, yaitu rentang 1 sampai 30
x_suhu['hangat'] = fuzz.trimf(x_suhu.universe, (20,30,40)) #rentang nilai batas pada inputan suhu hangat, yaitu rentang 20 sampai 40
x_suhu['panas']  = fuzz.trapmf(x_suhu.universe, (30, 40, 60, 70))#rentang nilai batas inputan suhu panas, yaitu rentang 30 sampai 60

#inputan kelembapan
x_kelembapan['kering'] = fuzz.trapmf(x_kelembapan.universe, (0, 0, 30, 50)) #rentang nilai batas pada inputan kelembapan kering, yaitu rentang 1 sampai 50
x_kelembapan['normal'] = fuzz.trimf(x_kelembapan.universe, (30, 50, 70))#rentang nilai batas pada inputan kelembapan normal, yaitu rentang 30 sampai 70
x_kelembapan['basah']  = fuzz.trapmf(x_kelembapan.universe, (50, 70, 100,100)) #rentang nilai batas pada inputan kelembapan basah, yaitu rentang 50 sampai 100

#output durasi pompa air
x_durasi['cepat']  = fuzz.trapmf(x_durasi.universe, (0, 0, 50, 100))#rentang nilai batas pada output durasi lama, yaitu rentang 1 sampai 100
x_durasi['sedang'] = fuzz.trimf(x_durasi.universe, (50, 100, 150)) #rentang nilai batas pada output durasi sedang, yaitu rentang 50 sampai 150
x_durasi['lama']   = fuzz.trapmf(x_durasi.universe, (100,150,250,250))#rentang nilai batas pada output durasi cepat, yaitu rentang 100 sampai 250
```

## RULE BASE
Pada kasus ini terdapat 9 rule base seperti pada tabel dibawah ini

| No |    SUHU   | KELEMBAPAN  | DURASI      |
|----|-----------|-------------|-------------|
| 1  |   Dingin  |   Kering    |   Lama      |
| 2  |   Dingin  |   Normal    |   Cukup     |
| 3  |   Dingin  |   Basah     |   Pendek    |
| 4  |   Hangat  |   Kering    |   Lama      |
| 5  |   Hangat  |   Normal    |   Cukup     |
| 6  |   Hangat  |   Basah     |   Pendek    |
| 7  |   Panas   |   Kering    |   Lama      |
| 8  |   Panas   |   Normal    |   Cukup     |
| 9  |   Panas   |   Basah     |   Pendek    |

```python
rule1 = ctr.Rule(x_suhu['dingin'] & x_kelembapan['kering'], x_durasi['lama'])
rule2 = ctr.Rule(x_suhu['dingin'] & x_kelembapan['normal'], x_durasi['sedang'])
rule3 = ctr.Rule(x_suhu['dingin'] & x_kelembapan['basah'], x_durasi['cepat'])

rule4 = ctr.Rule(x_suhu['hangat'] & x_kelembapan['kering'], x_durasi['lama'])
rule5 = ctr.Rule(x_suhu['hangat'] & x_kelembapan['normal'], x_durasi['sedang'])
rule6 = ctr.Rule(x_suhu['hangat'] & x_kelembapan['basah'], x_durasi['cepat'])

rule7 = ctr.Rule(x_suhu['panas'] & x_kelembapan['kering'],x_durasi['lama'])
rule8 = ctr.Rule(x_suhu['panas'] & x_kelembapan['normal'], x_durasi['sedang'])
rule9 = ctr.Rule(x_suhu['panas'] & x_kelembapan['basah'], x_durasi['cepat'])


rule1.view() #mencetak pola - pola rule
rule2.view()
rule3.view()
rule4.view()
rule5.view()
rule6.view()
rule7.view()
rule8.view()
rule9.view()
```

## 2. Fuzzy Inference

Dari beberapa metode dalam fuzzy inference system, metode yang digunakan adalah Metode Mamdani.

```python
#membuat kontrol sistem yang kondisi nya diambil berdasarkan aturan/rule yang telah dibuat
durasi_rule1 = ctr.ControlSystem([rule1, rule2, rule3, rule4, rule5, rule6, rule7, rule8, rule9])
```

```python
#membuat simulasi kontrol sistem yang kondisinya di ambil berdasarkan durasi_rule1/sistem kontrol yg telah kita buat
durasi = ctr.ControlSystemSimulation(durasi_rule1)
```

## 3. Defuzzifikasi
Defuzzifikasi merupakan pengubahan output fuzzy yang diperoleh dari mesin inferensi menjadi nilai tegas menggunakan fungsi keanggotaan yang sesuai dengan saat dilakukan fuzzyfikasi. Sekarang kita dapat mensimulasikan sistem kontrol hanya dengan menentukan input dan memanggil metode komputasi. 


```python
#membuat interupsi yang meminta inputan dari keyboard
Suhu            = input('Temperatur Suhu  :') #membuat inputan untuk nilai inputan sensor suhu          
Kelembapan      = input('Kelembapan Tanah :') #membuat inputan untuk nilai inputan sensor kelembapan tanah
```

<h1 align = "left"> <font color='red'> Contoh Kasus </h1> </font>
pada sebuah taman dipasang pompa air otomatis dengan atribut berupa dua buah sensor, yaitu sensor suhu dan sensor kelembapan. sensor suhu mendeteksi suhu lingkungan yaitu 22 derajat celcius, dan sensor kelembapan mendeteksi kelembapan tanah yaitu 35,maka bagaimanakah respon dari pompa air sebagai output dari kedua sensor tersebut?

1. $Input.SUHU[15] =
\left\{
    \begin{array}{ll}
       {30-22 \over 30-20} = 0.8 \\
       {22-20 \over 30-20} = 0.2 \\
    \end{array}
\right.$



2. $Input.KELEMBAPAN[40] =
\left\{
    \begin{array}{ll}
       {50-35 \over 50-30} = 0.75 \\
       {35-30 \over 50-30} = 0.25 \\
    \end{array}
\right.$

Mengambil nilai Minimal = (Min(0.8, 0.75), Min(0.8, 0.25), Min(0.2, 0.75), Min(0.2, 0.25) = (0.75, 0.25, 0.2, 0.2)

#### mencari hasil akhir dengan menggunkan metode COG (Centre Of Gravity)
$COG =
\left\{
    \begin{array}{ll}
       {( (50+100+150)x0.25 + (200)x0.75 \over (0.25+0.25+0.25+0.75)}= 150 \\
    \end{array}
\right.$

```python
durasi.input['x_suhu'] = float(Suhu) #merubah format data suhu menjadi float, dan memasukkannya kedalam kontrol sistem berdasarkan kondisi yang telah dibuat
durasi.input['x_kelembapan'] = float(Kelembapan) #merubah format data Kelembapan menjadi float, dan memasukkannya kedalam kontrol sistem berdasarkan kondisi yang telah dibuat

durasi.compute() #menghitung jumlahnya
```

```python
print(durasi.output['x_durasi']) #mencetak output durasi motor pompa berdasarkan kondisi yg telah di penuhi
x_durasi.view(sim=durasi)#membuat plot/grafik dari durasi motor pompa dengan kondisi berada pada durasi_rule yang telah terpenuhi
```
