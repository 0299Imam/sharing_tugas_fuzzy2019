```python
NilaiAkhir = ctrl.Antecedent(np.arange(0,101,1),'Nilai Akhir')
NilaiKeaktifan = ctrl.Antecedent(np.arange(0,101,1),'Nilai Keaktifan')
NilaiPrestasi = ctrl.Antecedent(np.arange(0,4,0.5),'Nilai Prestasi')
NilaiKeorganisasian = ctrl.Consequent(np.arange(75,101,1),'Nilai Keorganisasian')
```

Variabel Nilai Keaktifan dengan start 0 sampai 101 jarak 1 , antecedent adalah library untuk 'input'
    
Variabel Nilai Prestasi dengan start 0 sampai 5 jarak 0.5 , antecedent adalah library untuk 'input'

# # Respesentasi dengan grafik pada ' Nilai Akhir ' , ' Nilai Keaktifan ' dan ' Nilai Prestasi ' dapat digambarkan sebagai berikut :

```python
#( Input ) ' NIlai Akhir '
NilaiAkhir['kurang'] = fuzz.trapmf(NilaiAkhir.universe,[0,0,45,60])#membership Kurang dengan bentuk trapesium dengan nilai 0 - 60
NilaiAkhir['cukup'] = fuzz.trimf(NilaiAkhir.universe,[45,60,75])#membership cukup dengan bentuk segitiga dengan nilai 45 - 75
NilaiAkhir['baik'] = fuzz.trapmf(NilaiAkhir.universe,[60,75,100,100])#membership Baik dengan bentuk trapesium dengan nilai 60 - 100
NilaiAkhir.view()#untuk menampikan Nilai akhir

#( Input ) ' NIlai Keaktifan '
NilaiKeaktifan['kurang'] = fuzz.trapmf(NilaiKeaktifan.universe,[0,0,45,60])#membership kurang dengan bentuk trapesium dengan nilai 0 - 60
NilaiKeaktifan['cukup'] = fuzz.trimf(NilaiKeaktifan.universe,[45,60,75])#membership cukup dengan bentuk segitiga dengan nilai 45 - 75
NilaiKeaktifan['baik'] = fuzz.trapmf(NilaiKeaktifan.universe,[60,75,100,100])#membership baik dengan bentuk trapesium dengan nilai 60 - 100
NilaiKeaktifan.view()#untuk menampilkan nilai Keaktifan

#( Input ) ' NIlai Prestasi '
NilaiPrestasi['kurang'] = fuzz.trapmf(NilaiPrestasi.universe,[0,0,2,2.5])#membership kurang dengan bentuk trapesium dengan nilai 0 - 2,5
NilaiPrestasi['cukup'] = fuzz.trimf(NilaiPrestasi.universe,[2,2.5,3])#membership cukup dengan bentuk segitiga dengan nilai 2 - 3
NilaiPrestasi['baik'] = fuzz.trapmf(NilaiPrestasi.universe,[2.5,3,4,4])#membership Baik dengan bentuk trapesium dengan nilai 2,5 - 4
NilaiPrestasi.view()#untuk menampilkan nilai prestasi
```

$3.input.Nilai Prestasi[3,0] =
\left\{
    \begin {array}{ll}

         {3-2,5 \over 3-2,5} & \mbox{= } 1
     \end {array}
\right.$

### setelah derajat keanggotaan masing masing dihitung , Proses selanjutnya adalah menghitung defuzzifikasi dengan metode Center of grafity 

### Nilai Minimum pada input :

 Min ( 0, 0,33 ) <br> Min ( 0, 0,33 ) <br> Min ( 0, 0,33 ) <br> Min ( 0, 0,67 )

### Menentukan nilai dengan metode COG (Centre Of Gravity)
<center>$COG =
\left\{
    \begin {array}{ll}
         {(\ 85\ +\ 90\ )\ *\ 0,33\ +\ (\ 95\ +\ 100\ )\ *\ 0,67 \over (\ 0,33\ )\ +\ (\ 0,33\ )\ +\ (\ 0,67\ )\ +\ (\ 0,67\ )} 
     \end {array}
\right.$

<center>$COG =
\left\{
    \begin {array}{ll}
         { 57,75\ + \ 130.65 \over 2 } 
     \end {array}
\right.$

<center>$COG =
\left\{
    \begin {array}{ll}
         { 188.4 \over 2 } & \mbox{= } 94,2  
     \end {array}
\right.$
