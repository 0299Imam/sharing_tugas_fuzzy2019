## Fuzzyfikasi
### Menentukan variabel dan semesta pembicaraan
Pada tahap ini semua variabel input maupun output akan ditentukan nilai semesta pembicaraannya atau juga dapat dikatakan ini adalah penentuan dari batas dari setiap variabel 

```python
# Variable Input
Persentase_lahan_sakit = ctrl.Antecedent(np.arange(0, 51, 1), 'Persentase Lahan Sakit')
Bentuk_Wilayah = ctrl.Antecedent(np.arange(0,4,1), 'Bentuk Wilayah')
Derajat_Berat = ctrl.Antecedent(np.arange(0,4,1), 'Derajat Berat')
Ketebalan_lapis = ctrl.Antecedent(np.arange(0,70,1), 'Ketebalan  Lapis')
```

```python
# membuat membership function untuk variabel Persentasi Lahan Sakit
Persentase_lahan_sakit["Rendah"] = fuzz.trapmf(Persentase_lahan_sakit.universe,[0,0,10,30]) #membership function rendah, dalam bentuk trapesium dengan nilai 0 sampai 30
Persentase_lahan_sakit["Tinggi"] = fuzz.trapmf(Persentase_lahan_sakit.universe,[10,30,50,50]) #membership function tinggi, dalam bentuk trapesium dengan nilai 10 sampai 50
Persentase_lahan_sakit.view()
```

Persentase Lahan Sakit Rendah :

$Persentase Lahan Sakit Rendah [x] = 
\left\{
    \begin{array}{ll}
        & 1 \mbox{; } x \leq 10 \\
       {30-x \over 30-10} & \mbox{; } 10 < x > 30 \\
        & 0 \mbox{; } x \geq 30
    \end{array}
\right.$

Persentase Lahan Sakit Tinggi :

$Persentase Lahan Sakit Tinggi [x] = 
\left\{
    \begin{array}{ll}
        & 1 \mbox{; } x \geq 10 \\
       {x-10 \over 30-10} & \mbox{; } 10 < x > 30 \\
        & 0 \mbox{; } x \leq 30
    \end{array}
\right.$

```python
# membuat membership function untuk variabel Bentuk Wilayah
Bentuk_Wilayah['Rendah'] = fuzz.trapmf(Bentuk_Wilayah.universe, (0, 0, 1, 2)) #membership function rendah, dalam bentuk trapesium dengan nilai 0 sampai 2
Bentuk_Wilayah['Tinggi'] = fuzz.trapmf(Bentuk_Wilayah.universe, (1, 2, 3, 3)) #membership function tinggi, dalam bentuk trapesium dengan nilai 1 sampai 3
Bentuk_Wilayah.view()
```

```python
# membuat membership function untuk variabel Derajat Berat
Derajat_Berat['Rendah'] = fuzz.trapmf(Derajat_Berat.universe, (0, 0, 1, 2)) #membership function rendah, dalam bentuk trapesium dengan nilai 0 sampai 2
Derajat_Berat['Tinggi'] = fuzz.trapmf(Derajat_Berat.universe, (1, 2, 3, 3)) #membership function tinggi, dalam bentuk trapesium dengan nilai 1 sampai 3
Derajat_Berat.view()
```

```python
# membuat membership function untuk variabel Ketebalan Lapisan
Ketebalan_lapis['Rendah'] = fuzz.trapmf(Ketebalan_lapis.universe, (0, 0, 25, 30)) #membership function rendah, dalam bentuk trapesium dengan nilai 0 sampai 30
Ketebalan_lapis['Tinggi'] = fuzz.trapmf(Ketebalan_lapis.universe, (25, 30, 60, 60)) #membership function tinggi, dalam bentuk trapesium dengan nilai 25 sampai 60
Ketebalan_lapis.view()
```

```python
# Variable Input
hasil = ctrl.Consequent(np.arange(0,401,1), 'hasil')
```

```python
# membuat membership function untuk variabel keputusan dan hasil
hasil['Diterima'] = fuzz.trapmf(hasil.universe, (0,0,250,350))
hasil['Ditolak'] = fuzz.trapmf(hasil.universe, (250,350,400,400))

hasil.view()
```

Keputusan Rendah :

$Keputusan Rendah [x] = 
\left\{
    \begin{array}{ll}
        & 1 \mbox{; } x \leq 80 \\
       {100-x \over 100-80} & \mbox{; } 80 < x > 100 \\
        & 0 \mbox{; } x \geq 100
    \end{array}
\right.$

Keputusan Tinggi :

$Keputusan Tinggi [x] = 
\left\{
    \begin{array}{ll}
        & 1 \mbox{; } x \geq 100 \\
       {x-80 \over 100-80} & \mbox{; } 80 < x > 100 \\
        & 0 \mbox{; } x \leq 80
    \end{array}
\right.$

```python
# membuat aturan (rule base) untuk menentukan status pemilihan lahan
# karena input ada 2 dan setiap input memiliki 2 kondisi, maka saya membuat 16 aturan

rule1 = ctrl.Rule(Persentase_lahan_sakit['Rendah'] & Bentuk_Wilayah['Rendah'] & Derajat_Berat['Rendah'] & Ketebalan_lapis['Rendah'],(hasil['Diterima'] ))
rule2 = ctrl.Rule(Persentase_lahan_sakit['Rendah'] & Bentuk_Wilayah['Rendah'] & Derajat_Berat['Rendah'] & Ketebalan_lapis['Tinggi'],(hasil['Diterima'] ))
rule3 = ctrl.Rule(Persentase_lahan_sakit['Rendah'] & Bentuk_Wilayah['Rendah'] & Derajat_Berat['Tinggi'] & Ketebalan_lapis['Rendah'],(hasil['Diterima'] ))
rule4 = ctrl.Rule(Persentase_lahan_sakit['Rendah'] & Bentuk_Wilayah['Rendah'] & Derajat_Berat['Tinggi'] & Ketebalan_lapis['Tinggi'],(hasil['Diterima'] ))

rule5 = ctrl.Rule(Persentase_lahan_sakit['Rendah'] & Bentuk_Wilayah['Tinggi'] & Derajat_Berat['Rendah'] & Ketebalan_lapis['Rendah'],(hasil['Diterima'] ))
rule6 = ctrl.Rule(Persentase_lahan_sakit['Rendah'] & Bentuk_Wilayah['Tinggi'] & Derajat_Berat['Rendah'] & Ketebalan_lapis['Tinggi'],(hasil['Diterima'] ))
rule7 = ctrl.Rule(Persentase_lahan_sakit['Rendah'] & Bentuk_Wilayah['Tinggi'] & Derajat_Berat['Tinggi'] & Ketebalan_lapis['Rendah'],(hasil['Diterima'] ))
rule8 = ctrl.Rule(Persentase_lahan_sakit['Rendah'] & Bentuk_Wilayah['Tinggi'] & Derajat_Berat['Tinggi'] & Ketebalan_lapis['Tinggi'],(hasil['Ditolak'] ))

rule9 = ctrl.Rule(Persentase_lahan_sakit['Tinggi'] & Bentuk_Wilayah['Rendah'] & Derajat_Berat['Rendah'] & Ketebalan_lapis['Rendah'],(hasil['Diterima'] ))
rule10 = ctrl.Rule(Persentase_lahan_sakit['Tinggi'] & Bentuk_Wilayah['Rendah'] & Derajat_Berat['Rendah'] & Ketebalan_lapis['Tinggi'],(hasil['Ditolak'] ))
rule11 = ctrl.Rule(Persentase_lahan_sakit['Tinggi'] & Bentuk_Wilayah['Rendah'] & Derajat_Berat['Tinggi'] & Ketebalan_lapis['Rendah'],(hasil['Ditolak'] ))
rule12 = ctrl.Rule(Persentase_lahan_sakit['Tinggi'] & Bentuk_Wilayah['Rendah'] & Derajat_Berat['Tinggi'] & Ketebalan_lapis['Tinggi'],(hasil['Ditolak'] ))

rule13 = ctrl.Rule(Persentase_lahan_sakit['Tinggi'] & Bentuk_Wilayah['Tinggi'] & Derajat_Berat['Rendah'] & Ketebalan_lapis['Rendah'],(hasil['Ditolak'] ))
rule14 = ctrl.Rule(Persentase_lahan_sakit['Tinggi'] & Bentuk_Wilayah['Tinggi'] & Derajat_Berat['Rendah'] & Ketebalan_lapis['Tinggi'],(hasil['Ditolak'] ))
rule15 = ctrl.Rule(Persentase_lahan_sakit['Tinggi'] & Bentuk_Wilayah['Tinggi'] & Derajat_Berat['Tinggi'] & Ketebalan_lapis['Rendah'],(hasil['Ditolak'] ))
rule16 = ctrl.Rule(Persentase_lahan_sakit['Tinggi'] & Bentuk_Wilayah['Tinggi'] & Derajat_Berat['Tinggi'] & Ketebalan_lapis['Tinggi'],(hasil['Ditolak'] ))

```

```python
rule1.view() # mencetak gambar aturan
```

```python
#membuat kontrol sistem dimana kondisinya diambil berdasarkan aturan yang telah dibuat
tipping_ctrl = ctrl.ControlSystem([rule1, rule2, rule3, rule4, rule5,rule6, rule7, rule8, rule9, rule10, rule11, rule12, rule13, rule14, rule15, rule16])
```

```python
#membuat simulasi kontrol sistem dimana kondisinya diambil berdasarkan keputusan_rule1/sistem kontrol yang telah dibuat
tipping = ctrl.ControlSystemSimulation(tipping_ctrl)
```
