# _ -- FUZZYFIKASI --_

Fuzzyfikasi dalam metode logika fuzzy merupakan tahap awal yang harus dilakukan untuk menentukan variabel, himpunan fuzzy, semesta pembicaraan dan domain. Hal ini dilakukan karena setiap variabel memiliki nilai numerik yang terdiri dari batas awal dan batas akhir. Sehingga variabel baik input dan output harus ditentukan semesta pembicaraannya.

Apabila dijelaskan secara definisi, maka penjelasan dari setiap tahap penentuan dalam proses fuzzyfikasi dapat dijabarkan sebagai berikut ini:

- **Variabel** --> objek dalam penelitian yang memiliki nilai dari setiap varian yang dapat berubah - ubah.
- **Himpunan Fuzzy** --> grup atau kelompok yang mewakili keadaan tertentu dalam suatu variabel.
- **Semesta Pembicaraan** --> keseluruhan nilai/range sebagai batas operasi dalam suatu himpunan fuzzy.
- **Domain** --> keseluruhan nilai/range yang dalam lingkup semesta pembicaraan yang telah di tentukan, lalu dioperasikan dalam setiap himpunan fuzzy.

Dari variabel input dan variabel output yang telah di tentukan sebelumnya, maka langkah selanjutnya yaitu membuat aturan-aturan yang menghasilkan output berdasarkan input yang diberikan. Aturannya seperti berikut ini:

# _ -- Mekanisme Inferensi dan Defuzzyfikasi --_

### Mekanisme Inferensi

Pada tahap ini, aturan (rule base) yang telah ditentukan berdasarkan data yang telah didapatkan dinyatakan kebentuk IF ... Then 
dan dilanjutkan dengan menarik kesimpulan menjadi nilai Fuzzy output berdasarkan nilai fuzzy input.

1. IF **Detak Jantung** `LAMBAT` dan **Tekanan Darah** `RENDAH`,THEN **Tingkat Stres** `TENANG`
2. IF **Detak Jantung** `LAMBAT` dan **Tekanan Darah** `NORMAL`,THEN **Tingkat Stres** `TENANG`
3. IF **Detak Jantung** `LAMBAT` dan **Tekanan Darah** `TINGGI`,THEN **Tingkat Stres** `TEGANG`
4. IF **Detak Jantung** `NORMAL` dan **Tekanan Darah** `RENDAH`,THEN **Tingkat Stres** `TENANG`
5. IF **Detak Jantung** `NORMAL` dan **Tekanan Darah** `NORMAL`,THEN **Tingkat Stres** `TENANG`
6. IF **Detak Jantung** `NORMAL` dan **Tekanan Darah** `TINGGI`,THEN **Tingkat Stres** `TEGANG`
7. IF **Detak Jantung** `CEPAT`  dan **Tekanan Darah** `RENDAH`,THEN **Tingkat Stres**  `CEMAS`
8. IF **Detak Jantung** `CEPAT`  dan **Tekanan Darah** `NORMAL`,THEN **Tingkat Stres** `CEMAS`
9. IF **Detak Jantung** `CEPAT`  dan **Tekanan Darah** `TINGGI`,THEN **Tingkat Stres** `TEGANG`


Dari aturan yang telah dibuat diatas, diubah ke dalam bentuk koding berupa beberapa rule dimana rule ini di dapatkan dari :
jumlah himpunan fuzzy dipangkatkan dengan jumlah input = 3^2 = 9 rule. 

```python
rule1 = ctrl.Rule(jantung['lambat'] & tekanan_darah['rendah'], tingkatstres['tenang'])
rule2 = ctrl.Rule(jantung['lambat'] & tekanan_darah['normal'], tingkatstres['tenang'])
rule3 = ctrl.Rule(jantung['lambat'] & tekanan_darah['tinggi'], tingkatstres['tegang'])
rule4 = ctrl.Rule(jantung['normal'] & tekanan_darah['rendah'], tingkatstres['tenang'])
rule5 = ctrl.Rule(jantung['normal'] & tekanan_darah['normal'], tingkatstres['tenang'])
rule6 = ctrl.Rule(jantung['normal'] & tekanan_darah['tinggi'], tingkatstres['tegang'])
rule7 = ctrl.Rule(jantung['cepat'] & tekanan_darah['rendah'], tingkatstres['cemas'])
rule8 = ctrl.Rule(jantung['cepat'] & tekanan_darah['normal'], tingkatstres['cemas'])
rule9 = ctrl.Rule(jantung['cepat'] & tekanan_darah['tinggi'], tingkatstres['tegang'])
```

### Menggabungan semua rule untuk dijadikan sebagai control system
Semua rule yang telah ditentukan digabungkan untuk menentukan aksi kontrol apa yang akan dilakukan dalam merespon nilai masukkan yang nantinya diberikan. Kode yang dapat digunakan untuk melakukan penggabungan ini yaitu `ctrl.ControlSystem`.

```python
simulasi_stres = ctrl.ControlSystem([rule1, rule2, rule3, rule4, rule5, rule6, rule7, rule8, rule9])
```

# Membuat mode Simulasi

```python
prediksi_stres = ctrl.ControlSystemSimulation(simulasi_stres)
```

Apabila seseorang melakukan tes untuk mengetahui tingkat stress yang dimilikinya dimana setelah melakukan tes detak jantung diperoleh = 90 bpm. Sedangkan tekanan darah yang dimilikinya = 107 mmHg. Maka berapa persentase dan tingkat stres  (Tenang / Cemas / Tegang) yang dimiliki oleh orang tersebut ? .

```python
prediksi_stres.input['Detak Jantung / Heart Rate (HR)'] = 90
prediksi_stres.input['Tekanan Darah / Blood Pressure (BP)'] = 107
```

#### EVALUASI RULE BASE
    
Perhitungan yang dilakukan diatas menghasilkan rule base yang aktif beserta nilai numeriknya sebagai berikut ini:
1. Jika Detak Jantung adalah `Normal`(0,33) dan Tekanan darah adalah `Rendah` (0.6). Maka, tingkat stres adalah `Tenang`(0,33).
2. Jika Detak Jantung adalah `Normal`(0,33) dan Tekanan darah adalah `Normal` (0.2). Maka, tingkat stres adalah `Tenang` (0,2).
3. Jika Detak Jantung adalah `Cepat` (0,25) dan Tekanan darah adalah `Rendah` (0.6). Maka, tingkat stres adalah `Cemas` (0,25).
4. Jika Detak Jantung adalah `Cepat` (0,25) dan Tekanan darah adalah `Normal` (0.2). Maka, tingkat stres adalah `Cemas`(0,2).

Catatan : 
- Pada rule aktif 1, nilai input Jantung adalah `Normal`(0,33) dan Tekanan darah adalah `Rendah` (0.6), maka diambil nilai minimumnya yaitu 0,33 sebagai nilai output `Tenang`.
- Pada rule aktif 2, nilai input Jantung adalah `Normal`(0,33) dan Tekanan darah adalah `Normal` (0.2), maka diambil nilai minimumnya yaitu 0,2 sebagai nilai output `Tenang`.
- Pada rule aktif 3, nilai input Jantung adalah `Cepat`(0,25) dan Tekanan darah adalah `Rendah` (0.6), maka diambil nilai minimumnya yaitu 0,25 sebagai nilai output `Cemas`.
- Pada rule aktif 4, nilai input Jantung adalah `Cepat`(0,25) dan Tekanan darah adalah `Normal` (0.6), maka diambil nilai minimumnya yaitu 0,2 sebagai nilai output `Cemas`.


#### **AGREGASI** 
Dari evaluasi rule base di atas, dapat disimpulkan bahwa terdapat output yang sama dengan nilai yang berbeda. Pada tahap agregasi ini hasil output rule base If-Then digabungkan mejadi sebuah fuzzy set tunggal. Caranya yaitu dengan menggunakan nilai maksimal.
