## Fuzzyfikasi Suhu

```python
#rentan himpunan fuzzy input untuk Suhu (1-41)
suhu        = ctrl.Antecedent(np.arange (1,41,1),  'suhu')
#performa
suhu_x      = np.arange(1, 41, 1)
```

 ### Variabel Input Suhu Ruangan, yaitu : </strong>
 
     Suhu Ruangan(dibagi 5 bagian) :
        a.Dingin (1 - 10)
        b.Sejuk  (1 - 20)
        c.Normal (10 - 30)
        d.Hangat (20 - 40)
        e.Panas  (30 - 40)

### Nilai Linguistik pada Variabel Suhu ada 5, yaitu :

- Suhu Dingin
$$ Suhu Dingin [x] = 
\begin{cases}
    \frac{(10-x)}{(10-1)} & ; 1 <  x \leq\ 10 \\
    0 & ; x \geq 10
\end{cases}
$$

- Suhu Sejuk
$$ Suhu Sejuk [x] = 
\begin{cases}
    0 & ; x \leq 1\\
    \frac{(x-1)}{(10-1)} & ;1 < x \leq\ 10 \\
    \frac{(20-x)}{(20-10)} & ; 10 \leq  x < \ 20 \\
    0 & ; x \geq 20
\end{cases}
$$

- Suhu Normal
$$ Suhu Normal [x] = 
\begin{cases}
    0 & ; x \leq 10\\
    \frac{(x-10)}{(20-10)} & ;10 < x \leq\ 20 \\
    \frac{(30-x)}{(30-20)} & ; 20 \leq  x <\ 30 \\
    0 & ; x \geq 30
\end{cases}
$$

- Suhu Hangat
$$ Suhu Hangat [x] = 
\begin{cases}
    0 & ; x \leq 20\\
    \frac{(x-20)}{(30-20)} & ;20 < x \leq\ 30 \\
    \frac{(40-x)}{(40-30)} & ; 30 \leq  x < \ 40 \\
    0 & ; x \geq 30
\end{cases}
$$

- Suhu Panas
$$ Suhu Panas [x] = 
\begin{cases}
    \frac{(x-30)}{(40-30)} & ; 30 <  x \leq\ 40 \\
    0 & ; x \geq 40
\end{cases}
$$


## Membership Function
- Membership function yang digunakan yaitu fungsi segitiga atau triangular (trimf).

Membership function dari Suhu,yaitu :

```python
suhu['dingin']             = fuzz.trimf(suhu.universe, [1, 1, 10])
suhu['sejuk']              = fuzz.trimf(suhu.universe, [1, 10, 20])
suhu['normal']             = fuzz.trimf(suhu.universe, [10, 20, 30])
suhu['hangat']             = fuzz.trimf(suhu.universe, [20, 30, 40])
suhu['panas']              = fuzz.trimf(suhu.universe, [30, 40, 40])
```

## Inferencing
    Pada tahap inferencing, membuat aturan-aturan fuzzy yang dinyatakan dalam bentuk IF .... THEN ....
    
    Pada sistem ini berdasarkan jumlah fuzzifikasi dari variabel suhu yang berjumlah 5 dan fuzzifikasi dari variabel orang yang berjumlah 3, maka rule yang dibuat berjumlah 5 x 3 = 15 rule

    1. IF Suhu DINGIN dan Orang SEDIKIT THEN Kipas SANGAT LAMBAT dan Exhaust SANGAT PELAN
    2. IF Suhu DINGIN dan Orang SEDANG THEN Kipas SEDANG dan Exhaust SEDANG
    3. IF Suhu DINGIN dan Orang BANYAK THEN Kipas CEPAT dan Exhaust SEDANG

    4. IF Suhu SEJUK dan Orang SEDIKIT THEN Kipas LAMBAT dan Exhaust PELAN
    5. IF Suhu SEJUK dan Orang SEDANG THEN Kipas CEPAT dan Exhaust SEDANG
    6. IF Suhu SEJUK dan Orang BANYAK THEN Kipas CEPAT dan Exhaust CEPAT

    7. IF Suhu NORMAL dan Orang SEDIKIT THEN Kipas CEPAT  dan Exhaust SEDANG 
    8. IF Suhu NORMAL dan Orang SEDANG THEN Kipas SEDANG dan Exhaust CEPAT
    9. IF Suhu NORMAL dan Orang BANYAK THEN Kipas CEPAT dan Exhaust CEPAT

    10. IF Suhu HANGAT dan Orang SEDIKIT THEN Kipas CEPAT dan Exhaust SEDANG
    11. IF Suhu HANGAT dan Orang SEDANG THEN Kipas SANGAT CEPAT dan Exhaust CEPAT
    12. IF Suhu HANGAT dan Orang BANYAK THEN Kipas SANGAT CEPAT dan Exhaust SANGAT CEPAT

    13. IF Suhu PANAS dan Orang SEDIKIT THEN Kipas CEPAT dan Exhaust SEDANG
    14. IF Suhu PANAS dan Orang SEDANG THEN Kipas SANGAT CEPAT dan Exhaust CEPAT
    15. IF Suhu PANAS dan Orang BANYAK THEN Kipas SANGAT CEPAT dan Exhaust SANGAT CEPAT

```python
rule1 = ctrl.Rule(suhu['dingin'] & orang['sedikit'],([kipas["sangat lambat"],exhaust["sangat pelan"]]))
rule2 = ctrl.Rule(suhu['dingin'] & orang['sedang'],([kipas["sedang"],exhaust["sedang"]]))
rule3 = ctrl.Rule(suhu['dingin'] & orang['banyak'],([kipas["cepat"],exhaust["sedang"]]))

rule4 = ctrl.Rule(suhu['sejuk'] & orang['sedikit'],[kipas["lambat"],exhaust["pelan"]])
rule5 = ctrl.Rule(suhu['sejuk'] & orang['sedang'],[kipas["cepat"],exhaust["sedang"]])
rule6 = ctrl.Rule(suhu['sejuk'] & orang['banyak'],[kipas["cepat"],exhaust["cepat"]])

rule7 = ctrl.Rule(suhu['normal'] & orang['sedikit'],[kipas["cepat"],exhaust["sedang"]])
rule8 = ctrl.Rule(suhu['normal'] & orang['sedang'],[kipas["sedang"],exhaust["cepat"]])
rule9 = ctrl.Rule(suhu['normal'] & orang['banyak'],[kipas["cepat"],exhaust["cepat"]])

rule10 = ctrl.Rule(suhu['hangat'] & orang['sedikit'],[kipas["cepat"],exhaust["sedang"]])
rule11 = ctrl.Rule(suhu['hangat'] & orang['sedang'],[kipas["sangat cepat"],exhaust["cepat"]])
rule12 = ctrl.Rule(suhu['hangat'] & orang['banyak'],[kipas["sangat cepat"],exhaust["sangat cepat"]])

rule13 = ctrl.Rule(suhu['panas'] & orang['sedikit'],[kipas["cepat"],exhaust["sedang"]])
rule14 = ctrl.Rule(suhu['panas'] & orang['sedang'],[kipas["sangat cepat"],exhaust["cepat"]])
rule15 = ctrl.Rule(suhu['panas'] & orang['banyak'],[kipas["sangat cepat"],exhaust["sangat cepat"]])

rule1.view()
```

Mencari Nilai Rule 

    Rule4 = IF Suhu "Sejuk" dan Orang "Sedikit" Then Kipas "Lambat" dan Exhaust "Pelan"
          = MIN (0,5 , 1) 
          = 0.5

    Rule7 = IF Suhu "Normal" dan Orang "Sedikit" Then Kipas "Cepat" dan Exhaust "Sedang"
          = MIN (0,5 , 1) 
          = 0.5


## Rule Evaluation 

```python
keluaran_ctrl = ctrl.ControlSystem([rule1, rule2, rule3,rule4, rule5, rule6,rule7, rule8, rule9, rule10, 
                                  rule11, rule12, rule13, rule14, rule15])
```

## Defuzzyfikasi

```python
tingkat = ctrl.ControlSystemSimulation(keluaran_ctrl)
```

    Input suhu kita masukan 15 dan orang 10.
    Mencetak hasil Defuzzyfikasi.

```python
tingkat.input["suhu"] = 15
tingkat.input["orang"] = 10
tingkat.compute()
tingkat.output
```

Plotting fungsi keangggotaan Kipas setelah dimasukkan input nilai pada Kipas

```python
kipas.view(sim=tingkat)
```

Plotting fungsi keangggotaan Exhaust setelah dimasukkan input nilai pada Exhaust

```python
exhaust.view(sim=tingkat)
```
