&emsp;&emsp;&emsp;Logika Fuzzy adalah suatu cara yang tepat untuk memetakan suatu ruang input ke dalam suatu ruang output. Alasan menggunakan logika fuzzy antara lain konsep logika fuzzy lebih mudah dipahami selain itu logika fuzzy apabila terdapat data yang tidak tepat memiliki toleransi.

<br>
1.  basis aturan yang berisi aturan-aturan yang bersumber dari pakar
<br>2.  suatu mekanisme pengambilan keputusan dimana pakar mengambil keputusan dengan menerapkan pengetahuan yang dimiliki
<br>3.  proses fuzzikasi yang merubah besaran crisp kedalam besaran fuzzy

<font size="3"><font color="red"> Variabel Input  (Antecedents) :</font></font>
<br>&emsp;&emsp;&emsp;1. <font color="purple">Fungsi keanggotaan jarak EV-OV : Dekat1, Sedang1 dan Jauh1.</font>
<br>&emsp;&emsp;&emsp;2. <font color="green">Fungsi keanggotaan jarak EV-RV. : Dekat2 Dan Jauh2.</font>

<font size="3"><font color="red"> Variabel Output (Consequents) :</font></font>
<br>&emsp;&emsp;&emsp;1.<font color="brown"> Fungsi keanggotaan keluaran</font>
<font size="3"><font color="red"> Nilai Lingustik </font></font>
<br>&emsp;&emsp;&emsp;Kemudian terdapat pula nilai lingustik atau semesta pembicaraan yang merupakan keseluruhan nilai yang diperbolehkan untuk dioperasikan dalam suatu variabel fuzzy, dan terdapat juga Domain fuzzy yang merupakan keseluruhan nilai yang diizinkan dalam semesta pembicaraan dan boleh dioperasikan dalam suatu himpunan fuzzy. 
<br>&emsp;&emsp;&emsp;Penamaan kendaraan EV (ego vehicle) untuk menamai kendaraan yang dilengkapi HCAS, OV (oncoming vehicle) untuk kendaraan yang melaju ke arah EV pada lajur yang sama, dan RV (right vehicle) untuk kendaraan di jalur kanan dari EV

- OV disini sebagai variabel fungsi keanggotaan jarak EV-OV dengan nilai 0-350 dengan range 2
- RV sebagai variabel fungsi keanggotaan jarak EV-RV dengan nilai 0-201 dengan range 2
- KK sebagai variabel fungsi keanggotaan keluaran dengan nilai 0-3.5 dengan range 0.1


- shift.Antecedent sebagai input pada data fungsi keanggotaan
- shift.Consequent sebagai output pada data fungsi keluaran

- np.arange sebagai range setiap nilai

```python
import numpy as np
import skfuzzy as fuzz
from skfuzzy import control as shift
```

```python
OV = shift.Antecedent (np.arange (1,350,2),'Fungsi Keanggotaan Jarak EV-OV')
RV = shift.Antecedent (np.arange (1,201,2),'Fungsi Keanggotaan Jarak EV-RV')
KK = shift.Consequent (np.arange (0.1,3.5,0.1),'Fungsi Keanggotaan Keluaran')
```

<font size="5"><font color="blue"> Fuzzyfikasi</font></font>

&emsp;&emsp;&emsp;Fuzzyfikasi merupakan proses untuk mengubah input sistem yang mempunyai nilai tegas menjadi variabel linguistik munggunakan fungsi keanggotaan yang disimpan dalam basis pengetahuan fuzzy. Dalam hal ini, terdapat 3 variabel, yaitu 2 variabel input dan 1 variabel output.

&emsp;&emsp;&emsp;Berikut adalah persamaan garis pada Input:

<font color="purple">$µDekat(x) = 
\left\{
    \begin{array}{ll}
       1 & \mbox{; } x \leq 30 \\
       {80-x \over 30} & \mbox{; } 30 ≤ x ≤ 80 \\
       0 & \mbox{; } x \geq 80 \\ 
    \end{array}
\right.$

$µSedang(x) = 
\left\{
    \begin{array}{ll}
       0 & \mbox{; } x \leq 30 \\
       {x-30 \over 50} & \mbox{; } 30 ≤ x ≤ 80 \\
       {130-x \over 50} & \mbox{; } 80 ≤ x ≤ 130 \\  
    \end{array}
\right.$

$µJauh(x) = 
\left\{
    \begin{array}{ll}
       0 & \mbox{; } x \leq 80 \\
       {x-80 \over 50} & \mbox{; } 80 ≤ x ≤ 130 \\
       1 & \mbox{; } x \geq 130 \\  
    \end{array}
\right.$

```python
OV ['Dekat1'] = fuzz.trapmf(OV.universe,[1,1,30,80])
OV ['Sedang1'] = fuzz.trimf(OV.universe,[30,80,130])
OV ['Jauh1'] = fuzz.trapmf(OV.universe,[80,130,350,350])
OV.view()
```

&emsp;&emsp;&emsp;<font size="2"><font color="red">Disini kita ingin memunculkan grafik dari data yang sudah kita inputkan</font></font>
- OV sebagai variabel fungsi keanggotaan jarak EV-OV yang dimana terdapat 3 bagian pada tabel yaitu dekat1, sedang1, dan jauh1.
- fuzz.trapmf untuk menampilkan grafik garis berbentuk trapesium dengan nilai 0 samapai 80
- fuzz.trimf untuk menampilkan grafik garis berbentuk segitiga dengan nilai 30 sampai 130
- fuzz.trapmf untuk menampilkan grafik garis berbentuk trapesium dengan nilai 80 samapai 350
- view () untuk menampilkan hasil grafik

Berikut adalah persamaan garis pada Input:

<font color="green">$µDekat(x) = 
\left\{
    \begin{array}{ll}
       1 & \mbox{; } x \leq 60 \\
       {110-x \over 60} & \mbox{; } 60 ≤ x ≤ 110 \\
       0 & \mbox{; } x \geq 110 \\
    \end{array}
\right.$

$µJauh(x) = 
\left\{
    \begin{array}{ll}
       0 & \mbox{; } x \leq 60 \\
       {x-60 \over 50} & \mbox{; } 60 ≤ x ≤ 210 \\
       1 & \mbox{; } x \geq 210 \\
    \end{array}
\right.$


```python
RV ['Dekat2'] = fuzz.trapmf(RV.universe,[1,1,60,110])
RV ['Jauh2'] = fuzz.trapmf(RV.universe,[60,110,210,210])
RV.view()
```

&emsp;&emsp;&emsp;<font size="2"><font color="red">Disini kita ingin memunculkan grafik dari data yang sudah kita inputkan</font></font>
- RV sebagai variabel fungsi keanggotaan jarak EV-RV yang dimana terdapat 2 bagian pada tabel yaitu dekat2 dan jauh2.
- fuzz.trapmf untuk menampilkan grafik garis berbentuk trapesium dengan nilai 0 samapai 110
- fuzz.trapmf untuk menampilkan grafik garis berbentuk trapesium dengan nilai 60 samapai 200
- view () untuk menampilkan hasil grafik

&emsp;&emsp;&emsp;Berikut adalah persamaan garis pada Input:
    
<font color="brown">$µMaju(x) = 
\left\{
    \begin{array}{ll}
       1 & \mbox{; } x \leq 0.8 \\
       {1.0-x \over 0.8} & \mbox{; } 0.8 ≤ x ≤ 1.0 \\
       0 & \mbox{; } x \geq 1.0 \\
    \end{array}
\right.$

$µBerhenti(x) = 
\left\{
    \begin{array}{ll}
       0 & \mbox{; } x \leq 0.8,  x \geq 2.2 \\
       {x-0.8 \over 1.5} & \mbox{; } 0.8 ≤ x ≤ 1.5 \\
       {2.2-x \over 2.0} & \mbox{; } 2.0 ≤ x ≤ 2.2 \\
       1 & \mbox{; } x \leq 2.2 \\      
    \end{array}
\right.$

$µMenghindar(x) = 
\left\{
    \begin{array}{ll}
       0 & \mbox{; } x \leq 2.0 \\
       {x-2.0 \over 1.5} & \mbox{; } 2.0 ≤ x ≤ 3.5 \\ 
    \end{array}
\right.$

```python
KK ['Maju'] = fuzz.trapmf(KK.universe,[0.1,0.1,0.8,1])
KK ['Berhenti'] = fuzz.trapmf(KK.universe,[0.8,1,2,2.2])
KK ['Menghindar'] = fuzz.trapmf(KK.universe,[2,2.2,3.5,3.5])
KK.view()
```

&emsp;&emsp;&emsp;<font size="2"><font color="red">Disini kita ingin memunculkan grafik dari data yang sudah kita inputkan</font></font>
- KK sebagai variabel fungsi keanggotaan keluaran yang dimana terdapat 3 bagian pada tabel yaitu maju, berhenti dan menghindar.
- fuzz.trapmf untuk menampilkan grafik garis berbentuk trapesium dengan nilai 0 samapai 1
- fuzz.trapmf untuk menampilkan grafik garis berbentuk trapesium dengan nilai 0.8 samapai 2.2
- fuzz.trapmf untuk menampilkan grafik garis berbentuk trapesium dengan nilai 2 samapai 3
- view () untuk menampilkan hasil grafik

 <font size="5"><font color="blue">Rule Base</font></font>

| No | <font color="purple"> Fungsi Keanggotaan Jarak EV-OV</font></font> | <font color="green"> Fungsi Keanggotaan Jarak EV-RV</font></font>   | <font color="brown">Fungsi Keanggotaan Keluaran</font></font>  |
|----|---------------------------------|-----------------------------------|------------------------------|
| 1  |<center><font color="purple">Dekat1|<center><font color="green">Dekat2|<center><font color="brown">Berhenti  |
| 2  |<center><font color="purple">Dekat1|<center><font color="green">Jauh2 |<center><font color="brown">Menghindar|
| 3  |<center><font color="purple">Sedang1|<center><font color="green">Dekat2|<center><font color="brown">Berhenti |
| 4  |<center><font color="purple">Sedang1|<center><font color="green">Jauh2 |<center><font color="brown">Menghindar|  
| 5  |<center><font color="purple">Jauh 1|<center><font color="green">Dekat2 |<center><font color="brown">Maju     |
| 6  |<center><font color="purple">Jauh 1|<center><font color="green">Jauh2  |<center><font color="brown">Maju     |


```python
rule1 = shift.Rule (OV ['Dekat1']& RV ['Dekat2'], KK ['Berhenti'])
rule2 = shift.Rule (OV ['Dekat1']& RV ['Jauh2'] ,KK ['Menghindar'])
rule3 = shift.Rule (OV ['Sedang1']& RV ['Dekat2'] ,KK ['Berhenti'])
rule4 = shift.Rule (OV ['Sedang1']& RV ['Jauh2'] ,KK ['Menghindar'])
rule5 = shift.Rule (OV ['Jauh1']& RV ['Dekat2'] ,KK ['Maju'])
rule6 = shift.Rule (OV ['Jauh1']& RV ['Jauh2'] ,KK ['Maju'])
```

```python
rule1.view()
rule2.view()
rule3.view()
rule4.view()
rule5.view()
rule6.view()
```

&emsp;&emsp;&emsp;<font size="2"><font color="red">Program ini berfungsi untuk:</font></font> 
- Mencetak rule dari data yang sudah ada digrafik

 <font size="5"><font color="blue">Fuzzy Inference</font></font>

&emsp;&emsp;&emsp;Dari beberapa metode dalam fuzzy inference system, metode yang digunakan adalah Metode Mamdani.

```python
new = shift.ControlSystem ([rule1,rule2,rule3,rule4,rule5,rule6])
```

&emsp;&emsp;&emsp;<font size="2"><font color="red">Program ini berfungsi untuk:</font></font>
- membuat kontrol sistem yang kondisi nya diambil berdasarkan aturan/rule yang telah dibuat

```python
old = shift.ControlSystemSimulation (new)
```

&emsp;&emsp;&emsp;<font size="2"><font color="red">Program ini berfungsi untuk:</font></font>
- membuat simulasi kontrol sistem yang kondisinya di ambil berdasarkan sistem kontrol yg telah kita buat
