<strong> Jarak kendaraan mempunyai rentang nilai dari 1 meter - 21 meter </strong>

```python
jarak_kendaraan = ctrl.Antecedent(np.arange(1,21,1),'jarak_kendaraan')
```

# Fuzzifikasi Jarak kendaraan

Proses untuk mengubah variabel non numerik menjadi variabel non fuzzy (numerik) menjadi variabel fuzzy (linguistik)

<strong> Untuk Jarak Kendaraan , fungsi keanggotaan dibagi menjadi 3 bagian : </strong>
$$ Jarak Kendaraan $$
$$ Dekat = (1 - 10) meter$$
$$ Sedang = (5 - 15) meter$$
$$ Jauh = (10 - 20) meter$$

Dengan fungsi keanggotaan Jarak Kendaraan

\begin{equation}
  JarakKendaraan['dekat'] = \mu JarakKendaraanDekat(x)=\begin{cases}
    1, & \text{$1 \leq x \leq 5$}.\\
    (10 - x)/(10 - 5) , & \text{$5 < x < 10$}.\\
    0, & \text{$x\geq 10$}.
  \end{cases}
  \\
  JarakKendaraan['sedang'] = \mu JarakKendaraanSedang(x)=\begin{cases}
    0, & \text{$x \leq 5$ atau $x \geq 15$}.\\
    (x - 5)/(10 - 5) , & \text{$5 < x \leq 10$}.\\
    (15 - x)/(15 - 10) , & \text{$10 < x < 15$}.\\
  \end{cases}
  \\
  JarakKendaraan['jauh'] = \mu JarakKendaraanJauh(x)=\begin{cases}
    0, & \text{$x\leq 10$}.\\
    (x - 10)/(15 - 10) , & \text{$10 < x < 15$}.\\
    1, & \text{$15 \leq x \leq 20$}.\\
  \end{cases}
\end{equation}


```python
jarak_kendaraan['dekat'] = fuzz.trapmf(jarak_kendaraan.universe, [1, 1, 5, 10]) 
jarak_kendaraan['sedang'] = fuzz.trimf(jarak_kendaraan.universe, [5, 10, 15])
jarak_kendaraan['jauh'] = fuzz.trapmf(jarak_kendaraan.universe, [10, 15, 21, 21]) 
```

```python
jarak_kendaraan.view()
```

# Fuzzifikasi Output Kecepatan Mobil

<strong> Untuk Kecepatan Mobil , fungsi keanggotaan dibagi menjadi 3 bagian : </strong>
$$ Kecepatan Mobil $$
$$ Lambat = (1 - 60) meter$$
$$ Sedang = (40 - 80)meter$$
$$ Cepat = (60 - 101)meter$$

Dengan fungsi keanggotaan Kecepatan Mobil

\begin{equation}
  KecepatanMobil['lambat'] = \mu KecepatanMobilLambatt(x)=\begin{cases}
    1, & \text{$1 \leq x \leq 40$}.\\
    (60 - x)/(60 - 40) , & \text{$40 < x < 60$}.\\
    0, & \text{$x\geq 60$}.
  \end{cases}
  \\
  KecepatanMobil['sedang'] = \mu KecepatanMobilSedang(x)=\begin{cases}
    0, & \text{$x \leq 40$ atau $x \geq 80$}.\\
    (x - 40)/(60 - 40) , & \text{$40 < x \leq 60$}.\\
    (80 - x)/(80 - 60) , & \text{$60 < x < 80$}.\\
  \end{cases}
  \\
  KeceptanMobil['cepat'] = \mu KecepatanMobilCepat(x)=\begin{cases}
    0, & \text{$x\leq 60$}.\\
    (x - 60)/(80 - 60) , & \text{$7.5 < x < 10$}.\\
    1, & \text{$80 \leq x \leq 100$}.\\
  \end{cases}
\end{equation}


```python
Kecepatan_mobil['lambat'] = fuzz.trapmf(Kecepatan_mobil.universe, [1, 1, 40, 60]) 
Kecepatan_mobil['sedang'] = fuzz.trimf(Kecepatan_mobil.universe, [40, 60, 80]) 
Kecepatan_mobil['cepat'] = fuzz.trapmf(Kecepatan_mobil.universe, [60,80,100,101])
```

```python
Kecepatan_mobil.view()
```

<div class="alert alert-info">
<h2> <center> Mekanisme Inferensi

Pada tahap inferencing, membuat aturan-aturan fuzzy yang dinyatakan dalam bentuk IF .... THEN ....

1. IF Jarak Kendaraan DEKAT dan Jarak tikungan DEKAT THEN Kecepatan Mobil LAMBAT
2. IF Jarak Kendaraan DEKAT dan Jarak tikungan SEDANG THEN Kecepatan Mobil LAMBAT
3. IF Jarak Kendaraan DEKAT dan Jarak tikungan JAUH THEN Kecepatan Mobil LAMBAT

4. IF Jarak Kendaraan SEDANG dan Jarak tikungan DEKAT THEN Kecepatan Mobil LAMBAT
5. IF Jarak Kendaraan SEDANG dan Jarak tikungan SEDANG THEN Kecepatan Mobil SEDANG
6. IF Jarak Kendaraan SEDANG dan Jarak tikungan JAUH THEN Kecepatan Mobil SEDANG

7. IF Jarak Kendaraan JAUH dan Jarak tikungan DEKAT THEN Kecepatan Mobil SEDANG
8. IF Jarak Kendaraan JAUH dan Jarak tikungan SEDANG THEN Kecepatan Mobil SEDANG
9. IF Jarak Kendaraan JAUH dan Jarak tikungan JAUH THEN Kecepatan Mobil CEPAT

<div class="alert alert-info">
<h2> <center> Rules Base

```python
rule1 = ctrl.Rule(jarak_kendaraan['dekat'] & jarak_tikungan['dekat'], Kecepatan_mobil['lambat'])
rule2 = ctrl.Rule(jarak_kendaraan['dekat'] & jarak_tikungan['sedang'], Kecepatan_mobil['lambat'])
rule3 = ctrl.Rule(jarak_kendaraan['dekat'] & jarak_tikungan['jauh'], Kecepatan_mobil['lambat'])

rule4 = ctrl.Rule(jarak_kendaraan['sedang'] & jarak_tikungan['dekat'], Kecepatan_mobil['lambat'])
rule5 = ctrl.Rule(jarak_kendaraan['sedang'] & jarak_tikungan['sedang'], Kecepatan_mobil['sedang'])
rule6 = ctrl.Rule(jarak_kendaraan['sedang'] & jarak_tikungan['jauh'], Kecepatan_mobil['sedang'])

rule7 = ctrl.Rule(jarak_kendaraan['jauh'] & jarak_tikungan['dekat'], Kecepatan_mobil['sedang'])
rule8 = ctrl.Rule(jarak_kendaraan['jauh'] & jarak_tikungan['sedang'], Kecepatan_mobil['sedang'])
rule9 = ctrl.Rule(jarak_kendaraan['jauh'] & jarak_tikungan['jauh'], Kecepatan_mobil['cepat'])

rule1.view()
rule2.view()
rule3.view()
rule4.view()
rule5.view()
rule6.view()
rule7.view()
rule8.view()
rule9.view()
```

```python
Kecepatan_mobil_ctrl = ctrl.ControlSystem([rule1,rule2,rule3,rule4,rule5,rule6,rule7,rule8,rule9]) #Rule evaluation
```

<div class="alert alert-info">
<h2> <center> DEFUZZYFIKASI

```python
Kecepatanmobil = ctrl.ControlSystemSimulation(Kecepatan_mobil_ctrl)
```

```python
Kecepatanmobil.input['jarak_kendaraan'] = 6 #input jarak Kendaraan
Kecepatanmobil.input['jarak_tikungan'] = 9 #input jarak tikungan
```

<strong> Melakukan perhitungan Defuzzifikasi </strong>

```python
Kecepatanmobil.compute() 
```
