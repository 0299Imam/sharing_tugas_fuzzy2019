# SISTEM PENDUKUNG KEPUTUSAN UNTUK MENENTUKAN PEMBERIAN BONUS TAHUNAN MENGGUNAKAN METODE FUZZY LOGIC TIPE MAMDANI 


```python
import numpy as np
import skfuzzy as fuzz
import matplotlib.pyplot as plt
from skfuzzy import control as ctrl
```

Fuzzy Logic Untuk Menentukan Pemberian Bonus Tahunan. Dalam program ini terdapat input dan ouput. Input yang di gunakan ada 3 yaitu pendidikan,lama_bekerja,absensi. untuk ouput yang digunakan ada 1 yaitu bonus.Numpy merupakan singkatan dari numerical python, adalah module yang di khususkan untuk menganalisa data dan memelakukan secara numerik disimbolkan dalam bentuk array(susunan).

```python
pendidikan = np.arange(1,31,1)
lama_bekerja = np.arange(1,7,1)
absensi = np.arange(1,101,1)
```

Input Variabel : Variabel pendidikan Variabel lama_bekerja Variable absensi

Domain himpunan Fuzzy untuk variabel pendidikan: rendah = [1 17] sedang = [6 30] tinggi = [17 30]

Domain himpunan Fuzzy untuk variabel lama_bekerja: rendah = [1 3] sedang = [2 4] tinggi = [3 6]

Domain himpunan Fuzzy untuk variabel absensi: kurang = [1 50] sedang = [39 100]  tinggi =  [50 100]


```python
pendidikan = ctrl.Antecedent(np.arange(1,40,1), 'pendidikan')
lama_bekerja = ctrl.Antecedent(np.arange(1,7,1), 'lama_bekerja')
absensi = ctrl.Antecedent(np.arange(1,121,1), 'absensi')
bonus = ctrl.Consequent(np.arange(1,101,1),'bonus')
```

Output Variabel Bonus

Domain himpunan Fuzzy untuk variabel Bonus: tidak = [1 70] menerima = [70 100 ]


```python
pendidikan['rendah'] = fuzz.trapmf(pendidikan.universe, [1,1,6, 17])
pendidikan['sedang'] = fuzz.trimf(pendidikan.universe, [6, 17, 30])
pendidikan['tinggi'] = fuzz.trapmf(pendidikan.universe, [17, 30,40,40])

lama_bekerja['rendah'] = fuzz.trapmf(lama_bekerja.universe, [1, 1,2,3 ])
lama_bekerja['sedang'] = fuzz.trimf(lama_bekerja.universe, [2, 3, 4])
lama_bekerja['tinggi'] = fuzz.trapmf(lama_bekerja.universe, [3, 4, 6, 6])

absensi['kurang'] = fuzz.trapmf(absensi.universe, [1, 1, 39,50])
absensi['sedang'] = fuzz.trimf(absensi.universe, [39, 50, 100])
absensi['baik'] = fuzz.trapmf(absensi.universe, [50, 100, 120,120])

bonus['tidak'] = fuzz.trimf(bonus.universe, [1, 40, 70])
bonus['menerima'] = fuzz.trimf(bonus.universe, [70, 85, 100])
```

Rules

Mendefinisikan hubungan fuzzy antara variabel input dan output, dengan kata lain aturan-aturan ini dibentuk untuk menyatakan relasi antara input yang ada dengan output. Operator yang yang digunakan untuk menghubungkan antara dua input adalah operator AND, dan yang menetapkan antara input-output adalah IF-THEN. Dari 3 input dan 1 output fuzzy tersebut akan ditetapkan aturan-aturan berdasarkan nilai dari masing-masing input dan output. Dimana aturan-aturan tersebut merupakan penetapan dan akan menentukan status pemberian bonus. Dari 3 input dimana masing-masing setiap input memiliki 3 kondisi dan 1 output memiliki 2 kondisi , didapatkan 27 rule.


```python
rule1 = ctrl.Rule(pendidikan['rendah'] & lama_bekerja['rendah'] & absensi['kurang'],bonus['tidak'])
rule2 = ctrl.Rule(pendidikan['rendah'] & lama_bekerja['rendah'] & absensi['sedang'],bonus['tidak'])
rule3 = ctrl.Rule(pendidikan['rendah'] & lama_bekerja['rendah'] & absensi['baik'],bonus['tidak'])
rule4 = ctrl.Rule(pendidikan['rendah'] & lama_bekerja['sedang'] & absensi['kurang'],bonus['tidak'])
rule5 = ctrl.Rule(pendidikan['rendah'] & lama_bekerja['sedang'] & absensi['sedang'],bonus['tidak'])
rule6 = ctrl.Rule(pendidikan['rendah'] & lama_bekerja['sedang'] & absensi['baik'],bonus['menerima'])
rule7 = ctrl.Rule(pendidikan['rendah'] & lama_bekerja['tinggi'] & absensi['kurang'],bonus['tidak'])
rule8 = ctrl.Rule(pendidikan['rendah'] & lama_bekerja['tinggi'] & absensi['sedang'],bonus['menerima'])
rule9 = ctrl.Rule(pendidikan['rendah'] & lama_bekerja['tinggi'] & absensi['baik'],bonus['menerima'])
rule10 = ctrl.Rule(pendidikan['sedang'] & lama_bekerja['rendah'] & absensi['kurang'],bonus['tidak'])
rule11 = ctrl.Rule(pendidikan['sedang'] & lama_bekerja['rendah'] & absensi['sedang'],bonus['tidak'])
rule12 = ctrl.Rule(pendidikan['sedang'] & lama_bekerja['rendah'] & absensi['baik'],bonus['menerima'])
rule13 = ctrl.Rule(pendidikan['sedang'] & lama_bekerja['sedang'] & absensi['kurang'],bonus['tidak'])
rule14 = ctrl.Rule(pendidikan['sedang'] & lama_bekerja['sedang'] & absensi['sedang'],bonus['menerima'])
rule15 = ctrl.Rule(pendidikan['sedang'] & lama_bekerja['sedang'] & absensi['baik'],bonus['menerima'])
rule16 = ctrl.Rule(pendidikan['sedang'] & lama_bekerja['tinggi'] & absensi['kurang'],bonus['tidak'])
rule17 = ctrl.Rule(pendidikan['sedang'] & lama_bekerja['tinggi'] & absensi['sedang'],bonus['menerima'])
rule18 = ctrl.Rule(pendidikan['sedang'] & lama_bekerja['tinggi'] & absensi['baik'],bonus['menerima'])
rule19 = ctrl.Rule(pendidikan['tinggi'] & lama_bekerja['rendah'] & absensi['kurang'],bonus['tidak'])
rule20 = ctrl.Rule(pendidikan['tinggi'] & lama_bekerja['rendah'] & absensi['sedang'],bonus['menerima'])
rule21 = ctrl.Rule(pendidikan['tinggi'] & lama_bekerja['rendah'] & absensi['baik'],bonus['menerima'])
rule22 = ctrl.Rule(pendidikan['tinggi'] & lama_bekerja['sedang'] & absensi['kurang'],bonus['menerima'])
rule23 = ctrl.Rule(pendidikan['tinggi'] & lama_bekerja['sedang'] & absensi['sedang'],bonus['menerima'])
rule24 = ctrl.Rule(pendidikan['tinggi'] & lama_bekerja['sedang'] & absensi['baik'],bonus['menerima'])
rule25 = ctrl.Rule(pendidikan['tinggi'] & lama_bekerja['tinggi'] & absensi['kurang'],bonus['tidak'])
rule26 = ctrl.Rule(pendidikan['tinggi'] & lama_bekerja['tinggi'] & absensi['sedang'],bonus['menerima'])
rule27 = ctrl.Rule(pendidikan['tinggi'] & lama_bekerja['tinggi'] & absensi['baik'],bonus['menerima'])
```

```python
rule1.view()
```

# Rule Evaluation

```python
bonus_ctrl = ctrl.ControlSystem([rule1, rule2, rule3, rule4, rule5, rule6, rule7, rule8, rule9, rule10, rule11, rule12, rule13, rule14, rule15, rule16, rule17, rule18, rule19, rule20, rule21, rule22, rule23, rule24, rule25, rule26, rule27])
```

# Defuzzifikasi

```python
tingkat = ctrl.ControlSystemSimulation(bonus_ctrl)
```

untuk input kita masukkan

   -  input pendidikan = 30
   - input lama bekerja = 2
   - input absensi = 50



```python
tingkat.input['pendidikan'] = 30
tingkat.input['lama_bekerja'] = 2
tingkat.input['absensi'] = 50
```

Melakukan Perhitungan Defuzzifikasi

```python
tingkat.compute()
print (tingkat.output['bonus'])
```

Melakukan Plotting fungsi keanggotaan setalah memasukkan input pada input download

```python
bonus.view(sim=tingkat)
```
