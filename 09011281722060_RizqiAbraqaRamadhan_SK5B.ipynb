```python
#memanggil library yang akan kita gunakan
import numpy as np #import numpy, gunakan np untuk menggunakan library ini

import skfuzzy as fuzz #import library fuzzy, gunakan fuzz untuk menggunkaan library ini
#skfuzzy adalah library fuzzy logic pada python yang bekerja dengan numpy array

from skfuzzy import control as ctrl #panggil library control dari skfuzzy, gunakan ctrl untuk menggunakan library ini

import matplotlib.pyplot as plt #panggil library matplotlib, gunakan plt untuk menggunakan library ini
```

Terdapat 3 input dan 1 output yaitu :

Input :  1. Tes Tertulis      (0 - 100)
         2. Tes Keterampilan  (0 - 100)
         3. Tes Wawancara     (40 - 100)
         4. Tes Kesehtan      (50 - 80)
        
Output : 1. Hasil             (0 - 100)

```python
T = ctrl.Antecedent(np.arange(0, 101, 1),'Tes Tertulis') #Variabel T dengan start 0 sampai 101 jarak 1
K = ctrl.Antecedent(np.arange(0, 101, 1),'Tes Keterampilan') #Variabel K dengan start 0 sampai 101 jarak 1
W = ctrl.Antecedent(np.arange(30, 101, 1),'Tes Wawancara') #Variabel W dengan start 30 sampai 101 jarak 1
S = ctrl.Antecedent(np.arange(40, 81, 1),'Tes Kesehatan') #Variabel S dengan start 20 sampai 101 jarak 1
H = ctrl.Consequent(np.arange(0, 101, 1),'Hasil Seleksi') #Variabel H dengan start 0 sampai 101 jarak 1
```

## FUZZYFIKASI TES TERTULIS

$T.Lulus(x) = 
\left\{
    \begin{array}{ll}
       1 & \mbox{; } x \geq 75 \\
       {x-25 \over 75-25} & \mbox{; } 25 < x < 75 \\
       0 & \mbox{; } x \leq 25
    \end{array}
\right.$

$T.TidakLulus(x) = 
\left\{
    \begin{array}{ll}
       1 & \mbox{; } x \geq 25 \\
       {75-x \over 75-25} & \mbox{; } 25 < x < 75 \\
       0 & \mbox{; } x \leq 75
    \end{array}
\right.$

```python
#Input Tes Tertulis
T['Lulus'] = fuzz.trapmf(T.universe, [25, 75, 100, 100])
T['Tidak Lulus'] = fuzz.trapmf(T.universe, [0, 0, 25, 75])
T.view()
```

## RULES BASE

Pada tahap ini, akan dibuat aturan-aturan fuzzy yang dinyatakan dalam bentuk IF .... THEN ....

    Pada sistem ini berdasarkan jumlah fuzzifikasi dari : 1.variabel Tes Tertulis yang berjumlah 2.
                                                          2.variabel Tes Keterampilan yang berjumlah 2.
                                                          3.variabel Tes Wawancara berjumlah 2.
                                                          4.variabel Tes Kesehatan berjumlah 2.

maka rule yang dapat dibuat berjumlah 2x2x2x2 = 16 rule.

```python
rule1 = ctrl.Rule(T['Lulus'] & K['Lulus'] & W['Lulus'] & S['Lulus'], H['Diterima'])
rule2 = ctrl.Rule(T['Lulus'] & K['Lulus'] & W['Lulus'] & S['Tidak Lulus'], H['Tidak Diterima'])
rule3 = ctrl.Rule(T['Lulus'] & K['Lulus'] & W['Tidak Lulus'] & S['Lulus'], H['Tidak Diterima'])
rule4 = ctrl.Rule(T['Lulus'] & K['Lulus'] & W['Tidak Lulus'] & S['Tidak Lulus'], H['Tidak Diterima'])
rule5 = ctrl.Rule(T['Lulus'] & K['Tidak Lulus'] & W['Lulus'] & S['Lulus'], H['Diterima'])
rule6 = ctrl.Rule(T['Lulus'] & K['Tidak Lulus'] & W['Lulus'] & S['Tidak Lulus'], H['Tidak Diterima'])
rule7 = ctrl.Rule(T['Lulus'] & K['Tidak Lulus'] & W['Tidak Lulus'] & S['Lulus'], H['Tidak Diterima'])
rule8 = ctrl.Rule(T['Lulus'] & K['Tidak Lulus'] & W['Tidak Lulus'] & S['Tidak Lulus'], H['Tidak Diterima'])
rule9 = ctrl.Rule(T['Tidak Lulus'] & K['Lulus'] & W['Lulus'] & S['Lulus'], H['Diterima'])
rule10 = ctrl.Rule(T['Tidak Lulus'] & K['Lulus'] & W['Lulus'] & S['Tidak Lulus'], H['Tidak Diterima'])
rule11 = ctrl.Rule(T['Tidak Lulus'] & K['Lulus'] & W['Tidak Lulus'] & S['Lulus'], H['Tidak Diterima'])
rule12 = ctrl.Rule(T['Tidak Lulus'] & K['Lulus'] & W['Tidak Lulus'] & S['Tidak Lulus'], H['Tidak Diterima'])
rule13 = ctrl.Rule(T['Tidak Lulus'] & K['Tidak Lulus'] & W['Lulus'] & S['Lulus'], H['Diterima'])
rule14 = ctrl.Rule(T['Tidak Lulus'] & K['Tidak Lulus'] & W['Lulus'] & S['Tidak Lulus'], H['Tidak Diterima'])
rule15 = ctrl.Rule(T['Tidak Lulus'] & K['Tidak Lulus'] & W['Tidak Lulus'] & S['Lulus'], H['Tidak Diterima'])
rule16 = ctrl.Rule(T['Tidak Lulus'] & K['Tidak Lulus'] & W['Tidak Lulus'] & S['Tidak Lulus'], H['Tidak Diterima'])
```
