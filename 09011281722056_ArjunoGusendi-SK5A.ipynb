# 1. Import library fuzzy

```python
import numpy as np
import skfuzzy as fuzz
import matplotlib.pyplot as plt
from skfuzzy import control as ctrl
```

# 2. Fuzzyfikasi
### Menentukan variabel dan semesta pembicaraan

Pada langkah ini, setiap variabel input dan output akan ditentukan nilai semesta pembicaraannya atau penentuan batas nilai dari masing-masing variabel. Karena setiap variabel yang akan digunakan oleh proses fuzzy akan memiliki nilai numerik, dan memiliki batas awal dan batas akhir (crisp).

```python
#untuk plot terpisah
pot_x = np.arange(1,3000000,1)
unas_x = np.arange(1,10,0.05)
npre_x = np.arange(1,40,1)
kelayakan_x = np.arange(1,100,1)
```

```python
pot = ctrl.Antecedent(np.arange(1,3000000,1), 'pot')
unas = ctrl.Antecedent(np.arange(1,10,0.05), 'unas')
npre = ctrl.Antecedent(np.arange(1,40,1), 'npre')
kelayakan = ctrl.Consequent(np.arange(1,100,1),'kelayakan')
```

## A. Menentukkan Membership Function dan Nilai Linguistik untuk input

### Penghasilan Orang Tua (pot), terdiri atas 3 nilai linguistik, yaitu Rendah, Sedang, dan Tinggi. Maka fungsi keanggotaan dirumuskan sebagai berikut:

```python
pot['rendah'] = fuzz.trapmf(pot.universe, [1, 1, 500000,1000000])
pot['sedang'] = fuzz.trimf(pot.universe, [800000, 1000000, 1600000])
pot['tinggi'] = fuzz.trapmf(pot.universe, [1400000, 2500000, 3000000,3000000])
```

```python
pot_rendah = fuzz.trapmf(pot_x,[1,1,500000,1000000])
fig, (ax0) = plt.subplots(nrows=1,figsize=(8,6))
ax0.plot(pot_x,pot_rendah,'blue')
ax0.set_title('Rendah')
```

```python
pot_sedang = fuzz.trimf(pot_x,[800000,1000000,1600000])
fig, (ax0) = plt.subplots(nrows=1,figsize=(8,6))
ax0.plot(pot_x,pot_sedang,'orange')
ax0.set_title('sedang')
```

```python
pot_tinggi = fuzz.trapmf(pot_x,[1400000,2500000,3000000,3000000])
fig, (ax0) = plt.subplots(nrows=1,figsize=(8,6))
ax0.plot(pot_x,pot_tinggi,'green')
ax0.set_title('tinggi')
```

```python
pot.view()
```

# 3.  Penentuan Rule Base

```python
rule1 = ctrl.Rule(pot['rendah'] & unas['rendah'] & npre['rendah'],kelayakan['rendah'])
rule2 = ctrl.Rule(pot['rendah'] & unas['rendah'] & npre['sedang'],kelayakan['rendah'])
rule3 = ctrl.Rule(pot['rendah'] & unas['rendah'] & npre['tinggi'],kelayakan['rendah'])
rule4 = ctrl.Rule(pot['rendah'] & unas['sedang'] & npre['rendah'],kelayakan['rendah'])
rule5 = ctrl.Rule(pot['rendah'] & unas['sedang'] & npre['sedang'],kelayakan['tinggi'])
rule6 = ctrl.Rule(pot['rendah'] & unas['sedang'] & npre['tinggi'],kelayakan['tinggi'])
rule7 = ctrl.Rule(pot['rendah'] & unas['tinggi'] & npre['rendah'],kelayakan['tinggi'])
rule8 = ctrl.Rule(pot['rendah'] & unas['tinggi'] & npre['sedang'],kelayakan['tinggi'])
rule9 = ctrl.Rule(pot['rendah'] & unas['tinggi'] & npre['tinggi'],kelayakan['tinggi'])
rule10 = ctrl.Rule(pot['sedang'] & unas['rendah'] & npre['rendah'],kelayakan['rendah'])
rule11 = ctrl.Rule(pot['sedang'] & unas['rendah'] & npre['sedang'],kelayakan['rendah'])
rule12 = ctrl.Rule(pot['sedang'] & unas['rendah'] & npre['tinggi'],kelayakan['tinggi'])
rule13 = ctrl.Rule(pot['sedang'] & unas['sedang'] & npre['rendah'],kelayakan['tinggi'])
rule14 = ctrl.Rule(pot['sedang'] & unas['sedang'] & npre['sedang'],kelayakan['tinggi'])
rule15 = ctrl.Rule(pot['sedang'] & unas['sedang'] & npre['tinggi'],kelayakan['tinggi'])
rule16 = ctrl.Rule(pot['sedang'] & unas['tinggi'] & npre['rendah'],kelayakan['tinggi'])
rule17 = ctrl.Rule(pot['sedang'] & unas['tinggi'] & npre['sedang'],kelayakan['tinggi'])
rule18 = ctrl.Rule(pot['sedang'] & unas['tinggi'] & npre['tinggi'],kelayakan['tinggi'])
rule19 = ctrl.Rule(pot['tinggi'] & unas['rendah'] & npre['rendah'],kelayakan['rendah'])
rule20 = ctrl.Rule(pot['tinggi'] & unas['rendah'] & npre['sedang'],kelayakan['rendah'])
rule21 = ctrl.Rule(pot['tinggi'] & unas['rendah'] & npre['tinggi'],kelayakan['rendah'])
rule22 = ctrl.Rule(pot['tinggi'] & unas['sedang'] & npre['rendah'],kelayakan['rendah'])
rule23 = ctrl.Rule(pot['tinggi'] & unas['sedang'] & npre['sedang'],kelayakan['rendah'])
rule24 = ctrl.Rule(pot['tinggi'] & unas['sedang'] & npre['tinggi'],kelayakan['rendah'])
rule25 = ctrl.Rule(pot['tinggi'] & unas['tinggi'] & npre['rendah'],kelayakan['rendah'])
rule26 = ctrl.Rule(pot['tinggi'] & unas['tinggi'] & npre['sedang'],kelayakan['rendah'])
rule27 = ctrl.Rule(pot['tinggi'] & unas['tinggi'] & npre['tinggi'],kelayakan['tinggi'])
```

```python
rule1.view()
rule2.view()
rule3.view()
rule4.view()
rule5.view()
rule6.view()
rule7.view()
rule8.view()
rule9.view()
rule10.view()
rule11.view()
rule12.view()
rule13.view()
rule14.view()
rule15.view()
rule16.view()
rule17.view()
rule18.view()
rule19.view()
rule20.view()
rule21.view()
rule22.view()
rule23.view()
rule24.view()
rule25.view()
rule26.view()
rule27.view()
```

### Menggabungan semua rule untuk dijadikan sebagai control system

```python
beasiswa_ctrl =ctrl.ControlSystem([rule1,rule2,rule3,rule4,rule5,rule6,rule7,rule8,rule9,rule10,rule11,rule12,rule13,rule14,rule15,rule16,rule17,rule18,rule19,rule20,rule21,rule22,rule23,rule24,rule25,rule26,rule27])
```

# 5. Defuzzyfikasi

**PERHITUNGAN MANUAL MENGGUNAKAN METODE HITUNG LUAS SAMPLING**



**PERHITUNGAN KOMPUTASI**

### Menggunakan mode simulasi

```python
beasiswa = ctrl.ControlSystemSimulation(beasiswa_ctrl)
```

### Simulasi dengan memasukan nilai pada variabel input

   **disini saya mensimulasikan jika diinputkan nilai:**
  - Penghasilan Orang Tua = Rp.500000
  - Nilai Ujian Nasional = 8.5
  - Nilai Prestasi = 15

```python
beasiswa.input['pot'] = 500000
beasiswa.input['unas'] = 8.5
beasiswa.input['npre'] = 15
```

```python
layak_beasiswa = beasiswa.compute()
```

```python
layak_beasiswa = beasiswa.output['kelayakan']
print ("Perhitungan hasil komputasi = %.3f" % layak_beasiswa)
```

Perhitungan persentase **akurasi** antara perhitungan manual dengan perhitungan hasil komputasi

```python
z = 64.0220469
akurasi = layak_beasiswa*100/z
error = 100-akurasi
print ("Akurasi = %.3f" % akurasi + " %")
print ("Error = %.3f" % error  + " %")
```

```python
kelayakan.view(sim=beasiswa)
```

Setelah melihat perhitungan diatas , maka dapat disimpulkan bahwa kelayakan (X) penerimaan beasiswa Bidikmisi kepada pelajar/mahasiswa tersebut tergolong **TINGGI**
