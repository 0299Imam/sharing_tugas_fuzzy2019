### Menentukan variabel dan semesta pembicaraan

Pada langkah ini, setiap variabel input dan output akan ditentukan nilai semesta pembicaraannya atau penentuan batas nilai dari masing-masing variabel. Karena setiap variabel yang akan digunakan oleh proses fuzzy akan memiliki nilai numerik, dan memiliki batas awal dan batas akhir (crisp).

```python
#untuk plot terpisah
pot_x = np.arange(1,3000000,1)
unas_x = np.arange(1,10,0.05)
npre_x = np.arange(1,40,1)
kelayakan_x = np.arange(1,100,1)
```

```python
pot = ctrl.Antecedent(np.arange(1,3000000,1), 'pot')
unas = ctrl.Antecedent(np.arange(1,10,0.05), 'unas')
npre = ctrl.Antecedent(np.arange(1,40,1), 'npre')
kelayakan = ctrl.Consequent(np.arange(1,100,1),'kelayakan')
```

## A. Menentukkan Membership Function dan Nilai Linguistik untuk input

```python
pot['rendah'] = fuzz.trapmf(pot.universe, [1, 1, 500000,1000000])
pot['sedang'] = fuzz.trimf(pot.universe, [800000, 1000000, 1600000])
pot['tinggi'] = fuzz.trapmf(pot.universe, [1400000, 2500000, 3000000,3000000])
```
