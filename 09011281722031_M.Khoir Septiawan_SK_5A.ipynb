## Mengimport library untuk fuzzy

```python
import numpy as np
import skfuzzy as fuzz
from skfuzzy import control as shift
```

## Pembentukan himpunan fuzzy
Pada kasus ini ada 3 variabel yang akan dimodelkan, yaitu:
- Permintaan terdiri atas 2 himpunan fuzzy, yaitu (Turun dan Naik)
- Persediaan terdiri atas 2 himpunan fuzzy, yaitu (Sedikit dan Banyak)
- Produksi terdiri atas 2 himpunan fuzzy, yaitu (Berkurang dan Bertambah)

<b><u>INPUT</u><br/>
Permintaan</b>
- Turun (1,1,7965,11090)
- Naik (7965,11090,11090)

<b>Persediaan</b><br/>
- Sedikit (1,1,190,1250)
- Banyak (190,1250,1250)

<b><u>OUTPUT</u><br/>
Produksi</b></br>
- Berkurang (1,1,8150,12350)
- Bertambah (8150,12350,12350)

program dibawah ini untuk menginisiasikan range variable dari setiap input (pmt,psd) dan outputnya (pro)

```python
pmt = shift.Antecedent (np.arange (0,11090,1),'permintaan roti')
psd = shift.Antecedent (np.arange (0,1250,1),'persediaan roti')
pro = shift.Consequent (np.arange (0,12350,1),'produksi roti')
```

## Nilai Linguistik
a.Nilai Linguistik pada Variabel Permintaan ada 2, yaitu:

- Permintaan Turun
$$permintaan[μpmtTurun](x) =
\left\{
 \begin{array}{ll}
 1 & \mbox{;}x\leq 7965 \\
 {11090-x \over 11090-7965} & \mbox{;} 7965 \leq x \leq 11090 \\
 0 & \mbox{;} x \geq 11090
 \end {array}
\right.$$

- Permintaan Naik
$$permintaan[μpmtNaik] (y)=
\left\{
 \begin{array}{ll}
 0 & \mbox{;}x\leq 7965 \\
 {x-7965 \over 11090-7965} & \mbox{;} 7965 \leq x \leq 11090 \\
 1 & \mbox{;} x \geq 11090
 \end {array}
\right.$$

## Membership Function 
- Membership function yang digunakan yaitu fungsi trapesium (trapmf) dan segitiga (trimf). 

```python
pmt ['Turun'] = fuzz.trapmf(pmt.universe, [0,0,7965,11090])
pmt ['Naik'] = fuzz.trimf(pmt.universe, [7965,11090,11090])
pmt.view()
```

b. Nilai Linguistik pada Variabel Persediaan ada 2, yaitu:

- Persediaan Sedikit
$$persediaan[μpsdSedikit](y) =
\left\{
 \begin{array}{ll}
 1 & \mbox{;}x\leq 150 \\
 {1250-y \over 1250-150} & \mbox{;} 150 \leq x \leq 1250 \\
 0 & \mbox{;} x \geq 1250
 \end {array}
\right.$$

- Persediaan Banyak
$$persediaan[μpsdBanyak](y) =
\left\{
 \begin{array}{ll}
 0 & \mbox{;}x\leq 150 \\
 {x-150 \over 1250-150} & \mbox{;} 150 \leq x \leq 1250 \\
 1 & \mbox{;} x \geq 1250
 \end {array}
\right.$$

```python
psd ['Sedikit'] = fuzz.trapmf(psd.universe, [0,0,190,1250])
psd ['Banyak'] = fuzz.trimf(psd.universe, [190,1250,1250])
psd.view()
```

c. Nilai Linguistik pada Variabel Produksi ada 2, yaitu:

- Produksi Berkurang
$$produksi[μproBerkurang](z) =
\left\{
 \begin{array}{ll}
 1 & \mbox{;}x\leq 8150 \\
 {12350-x \over 12350-8150} & \mbox{;} 8150 \leq x \leq 12350 \\
 0 & \mbox{;} x \geq 12350
 \end {array}
\right.$$

- Produksi Bertambah
$$produksi[μproBertambah](z) =
\left\{
 \begin{array}{ll}
 0 & \mbox{;}x\leq 8150 \\
 {x-8150 \over 1250-8150} & \mbox{;} 8150 \leq x \leq 12350 \\
 1 & \mbox{;} x \geq 12350
 \end {array}
\right.$$

```python
pro ['Berkurang'] = fuzz.trapmf(pro.universe, [0,0,8150,12350])
pro ['Bertambah'] = fuzz.trimf(pro.universe, [8150,12350,12350])
pro.view()
```

## Aplikasi fungsi implikasi
- Fuzzy Rules

Disini saya menggunakan program untuk memasukan beberapa rule diantaranya :<br/>
R1 = JIKA permintaan TURUN dan persediaan BANYAK, MAKA produksi barang BERKURANG<br/>
R2 = JIKA permintaan TURUN dan persediaan SEDIKIT, MAKA produksi barang BERKURANG<br/>
R3 = JIKA permintaan NAIK dan persediaan BANYAK, MAKA produksi barang BERTAMBAH<br/>
R4 = JIKA permintaan NAIK dan persediaan SEDIKIT, MAKA produksi barang BERTAMBAH<br/>

```python
Rule1 = shift.Rule (pmt ['Turun']& psd ['Banyak'], pro['Berkurang'])
Rule2 = shift.Rule (pmt ['Turun']& psd ['Sedikit'], pro['Berkurang'])
Rule3 = shift.Rule (pmt ['Naik']& psd ['Banyak'], pro['Bertambah'])
Rule4 = shift.Rule (pmt ['Naik']& psd ['Sedikit'], pro['Bertambah'])
```

## Rule Evaluation
program dibawah ini adalah sebuah variable untuk memasukkan semua rule nya

```python
new = shift.ControlSystem ([Rule1,Rule2,Rule3,Rule4])
```

## Defuzzyfikasi
program dibawah ini untuk mensimulasikan variable dari new

```python
sk = shift.ControlSystemSimulation (new)
```
