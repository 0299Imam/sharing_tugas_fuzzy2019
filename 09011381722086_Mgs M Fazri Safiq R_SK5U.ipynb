Pada dasarnya Fuzzy logic merupakan logika bernilai banyak/ multivalued logic yang mampu mendefinisikan nilai diantara keadaan yang konvensional seperti benar atau salah, ya atau tidak, putih atau hitam dan lain-lain. Penalaran Logika Fuzzy memnyediakan cara untuk memahami kinerja system dengan cara menilai input dan output system dari hasil pengamatan.

Model Fuzzy Logic di bawah ini menggunakan metode Fuzzy Logic Mamdani

Metode ini dikemukakan oleh Mamdani dan Assilian (1975) untuk mengendalikan mesin uap dan mendidihkan berdasarkan sintesis himpunan kendali aturan linguistic dengan percobaan operator seorang manusia.
Metode sistem penalaran fuzzy Mamdani berdasarkan pada makalah Dr. Lofti Zadeh (1973) tentang algoritma fuzzy untuk sistem yang kompleks dan proses pengambilan keputusan.  Kelebihan metode Mamdani dibandingkan metode sistem penalaran fuzzy yang lain, diantaranya adalah karena bersifat intuitif, mencakup bidang yang luas, dan sesuai dengan proses input informasi manusia. 

Dalam Penilitian Efektifitas Elearning ini menggunakan metode FIS Mamdani, secara Umum terdapat 3 tahapan yaitu:
* [Fuzzyfikasi](#fikasi)
Yaitu suatu Proses fungsi pemetaan nilai inputkeanggotaan Fuzzy (Derajat Keanggotaan).
* [Inference](#ference)
Yaitu sesuatu yang dilakukan dengan aturan yang berhubungan if-then ya terait dengan beberapa in-putan variabel dan outputnya (Rule)
* [Defuzzifikasi](#de)
Yaitu suatu konversi fuzzy internal variabel outputke nilai tegas yang dapat secara aktual digunakan.Metode perhitungan centroid secara umumdigunakan dalam defuzzifikasi.

```python
import numpy as np
import skfuzzy as fuzz
from skfuzzy import control as ctrl
import matplotlib.pyplot as plt
```

- Import numpy bermaksud untuk menampung data yang akan diproses
- import skfuzzy berisi modul yang memproses logika fuzzy

```python
PEU = ctrl.Antecedent(np.arange(0, 100, 1), 'PEU')
PU = ctrl.Antecedent(np.arange(0, 100, 1), 'PU')
TEE = ctrl.Consequent(np.arange(0, 100, 1), 'TEE')
```

Penerapan e-learning dalam penerapan didunia pendidikan berdasarkan pengaruh faktor persepsi pemanfaatan (perceived usefulness) dan persepsi kemudahan penggunaan(perceived ease of use). Faktor Perceived Ease of Use merupakan variabel eksogen penelitian yang terdiri atas indikator yaitu: kemudahan berkomunikasi, mudah dipahami dan mudah digunakan. Sedangkan variabel yang ter-masuk ke dalam variabel endogen dalam penelitian ini adalah Perceived Usefullness yang terdiri indikator yang mampu meningkatkan produk-tivitas, kinerja, evaluasi serta efisiensi

Variabel input dan pada himpunan fuzzy output dibagi menjadi satu atau lebih

Input: Perceived Ease of Use (PEU)
- Sangat rendah (0,0,10,30)
- Rendah (10,30,50)
- Sedang (30,50,70)
- Tinggi (50,70,90)
- Sangat tinggi (70,90,100,100)

Input: Perceived Usefulness (PU)
- Sangat rendah (0,0,10,30)
- Rendah (10,30,50)
- Sedang (30,50,70)
- Tinggi (50,70,90)
- Sangat tinggi (70,90,100,100)

Output: Tingkat Efektivitas E-learning (TEE)
- Sangat rendah (0,0,10,30)
- Rendah (10,30,50)
- Cukup (30,50,70)
- Tinggi (50,70,90)
- Sangat tinggi (70,90,100,100)


# Fuzzyfikasi  <a class="anchor" id="fikasi"></a>

Untuk Perceived Ease of Use (PEU), fungsi keanggotaan dibagi menjadi 5 bagian :
$$  $$
$$ Sangat rendah = 0 - 30$$
$$ Rendah = 10 - 50$$
$$ Sedang = 30 - 70$$
$$ Tinggi = 50 - 90$$
$$ Sangat tinggi = 70 - 100$$

```python
PEU['sangat_rendah'] = fuzz.trapmf(PEU.universe, [0,0, 10, 30])
PEU['rendah'] = fuzz.trimf(PEU.universe, [10, 30, 50])
PEU['sedang'] = fuzz.trimf(PEU.universe, [30, 50, 70])
PEU['tinggi'] = fuzz.trimf(PEU.universe, [50, 70, 90])
PEU['sangat_tinggi'] = fuzz.trapmf(PEU.universe, [70, 90, 100, 100])
```

\begin{equation}
  PEU['sangat_rendah'] = \mu sangatrendah(x)=\begin{cases}
    1, & \text{$0\leq x\leq10$}.\\
    (30 - x)/(30 - 10) , & \text{$10 \leq x \leq 30$}.\\
    0, & \text{$x\geq 30$}.
  \end{cases}
  \\
  PEU['rendah'] = \mu rendah(x)=\begin{cases}
    0, & \text{$x\leq 10$ atau $x\geq50$}.\\
    (x - 10)/(30 - 10) , & \text{$10 \leq x \leq 30$}.\\
    (50 - x)/(50 - 30) , & \text{$30 \leq x \leq 50$}.\\
  \end{cases}
  \\
   PEU['sedang'] = \mu sedang(x)=\begin{cases}
    0, & \text{$x\leq 30$ atau $x\geq70$}.\\
    (x - 30)/(50 - 30) , & \text{$30 \leq x \leq 50$}.\\
    (70 - x)/(70 - 50) , & \text{$50 \leq x \leq 70$}.\\
  \end{cases}
  \\
   PEU['tinggi'] = \mu tinggi(x)=\begin{cases}
    0, & \text{$x\leq 50$ atau $x\geq90$}.\\
    (x - 50)/(70 - 50) , & \text{$50 \leq x \leq 70$}.\\
    (90 - x)/(90 - 70) , & \text{$70 \leq x \leq 90$}.\\
  \end{cases}
  \\
   PEU['sangat_tinggi'] = \mu sangattinggi(x)=\begin{cases}
    0, & \text{$x\leq 70$}.\\
    (x - 70)/(90 - 70) , & \text{$70 \leq x \leq 90$}.\\
    1, & \text{$90 \leq x \leq 100$}.
  \end{cases}
\end{equation}


```python
sangat_rendah = fuzz.trapmf(PEU.universe,[0,0, 10, 30])
fig, (ax0) = plt.subplots(nrows=1,figsize=(6,6))
ax0.plot(PEU.universe,sangat_rendah,'blue')
ax0.set_title('sangat_rendah')
```

# Inferencing  <a class="anchor" id="rencing"></a>


- IF PEU sangat_rendah AND PU sangat_rendah THEN TEE sangat_rendah
- IF PEU sangat_rendah AND PU rendah  THEN TEE rendah
- IF PEU sangat_rendah AND PU sedang THEN TEE rendah 
- IF PEU sangat_rendah AND PU tinggi  THEN TEE cukup
- IF PEU sangat_rendah AND PU sangat_tinggi THEN TEE cukup
- IF PEU rendah AND PU sangat_rendah  THEN TEE rendah
- IF PEU rendah AND PU rendah THEN TEE rendah
- IF PEU rendah AND PU sedang THEN TEE cukup
- IF PEU rendah AND PU tinggi THEN TEE cukup
- IF PEU rendah AND PU sangat_tinggi THEN TEE tinggi
- IF PEU sedang AND PU sangat_rendah THEN TEE rendah
- IF PEU sedang AND PU rendah THEN TEE rendah
- IF PEU sedang AND PU sedang THEN TEE cukup 
- IF PEU sedang AND PU tinggi THEN TEE cukup 
- IF PEU sedang AND PU sangat_tinggi THEN TEE tinggi
- IF PEU tinggi AND PU sangat_rendah THEN TEE cukup
- IF PEU tinggi AND PU rendah THEN TEE cukup
- IF PEU tinggi AND PU sedang THEN TEE tinggi
- IF PEU tinggi AND PU tinggi THEN TEE sangat_tinggi
- IF PEU tinggi AND PU sangat_tinggi THEN TEE tinggi
- IF PEU sangat_tinggi AND PU sangat_rendah THEN TEE cukup 
- IF PEU sangat_tinggi AND PU rendah THEN TEE tinggi
- IF PEU sangat_tinggi AND PU sedang THEN TEE tinggi 
- IF PEU sangat_tinggi AND PU tinggi THEN TEE sangat_tinggi
- IF PEU sangat_tinggi AND PU sangat_tinggi THEN TEE sangat_tinggi

```python
rule1 = ctrl.Rule(PEU['sangat_rendah'] & PU['sangat_rendah'], TEE['sangat_rendah'])
rule2 = ctrl.Rule(PEU['sangat_rendah'] & PU['rendah'], TEE['rendah'])
rule3 = ctrl.Rule(PEU['sangat_rendah'] & PU['sedang'], TEE['rendah'])
rule4 = ctrl.Rule(PEU['sangat_rendah'] & PU['tinggi'], TEE['cukup'])
rule5 = ctrl.Rule(PEU['sangat_rendah'] & PU['sangat_tinggi'], TEE['cukup'])
rule6 = ctrl.Rule(PEU['rendah'] & PU['sangat_rendah'], TEE['rendah'])
rule7 = ctrl.Rule(PEU['rendah'] & PU['rendah'], TEE['rendah'])
rule8 = ctrl.Rule(PEU['rendah'] & PU['sedang'], TEE['cukup'])
rule9 = ctrl.Rule(PEU['rendah'] & PU['tinggi'], TEE['cukup'])
rule10 = ctrl.Rule(PEU['rendah'] & PU['sangat_tinggi'], TEE['tinggi'])
rule11 = ctrl.Rule(PEU['sedang'] & PU['sangat_rendah'], TEE['rendah'])
rule12 = ctrl.Rule(PEU['sedang'] & PU['rendah'], TEE['rendah'])
rule13 = ctrl.Rule(PEU['sedang'] & PU['sedang'], TEE['cukup'])
rule14 = ctrl.Rule(PEU['sedang'] & PU['tinggi'], TEE['cukup'])
rule15 = ctrl.Rule(PEU['sedang'] & PU['sangat_tinggi'], TEE['tinggi'])
rule16 = ctrl.Rule(PEU['tinggi'] & PU['sangat_rendah'], TEE['cukup'])
rule17 = ctrl.Rule(PEU['tinggi'] & PU['rendah'], TEE['cukup'])
rule18 = ctrl.Rule(PEU['tinggi'] & PU['sedang'], TEE['tinggi'])
rule19 = ctrl.Rule(PEU['tinggi'] & PU['tinggi'], TEE['sangat_tinggi'])
rule20 = ctrl.Rule(PEU['tinggi'] & PU['sangat_tinggi'], TEE['tinggi'])
rule21 = ctrl.Rule(PEU['sangat_tinggi'] & PU['sangat_rendah'], TEE['cukup'])
rule22 = ctrl.Rule(PEU['sangat_tinggi'] & PU['rendah'], TEE['tinggi'])
rule23 = ctrl.Rule(PEU['sangat_tinggi'] & PU['sedang'], TEE['tinggi'])
rule24 = ctrl.Rule(PEU['sangat_tinggi'] & PU['tinggi'], TEE['sangat_tinggi'])
rule25 = ctrl.Rule(PEU['sangat_tinggi'] & PU['sangat_tinggi'], TEE['sangat_tinggi'])
```
